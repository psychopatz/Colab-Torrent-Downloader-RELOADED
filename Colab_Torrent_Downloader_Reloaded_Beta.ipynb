```python
#@title ‚ÜêInitialize Downloader{ run: "auto" }
import os
import time
import shutil
import datetime
import ipywidgets as widgets
import pickle
import pytz

from datetime import datetime
from google.colab import files
from google.colab import output
from threading import Thread
from IPython.display import display, clear_output

Downloads = "/content/drive/MyDrive/Downloads" #@param {type:"string"}
SaveDirectory = Downloads + '/MyTorrents' 
Timezone = 'Asia/Manila' #@param ["Africa/Abidjan","Africa/Accra","Africa/Addis_Ababa","Africa/Algiers","Africa/Asmara","Africa/Bamako","Africa/Bangui","Africa/Banjul","Africa/Bissau","Africa/Blantyre","Africa/Brazzaville","Africa/Bujumbura","Africa/Cairo","Africa/Casablanca","Africa/Ceuta","Africa/Conakry","Africa/Dakar","Africa/Dar_es_Salaam","Africa/Djibouti","Africa/Douala","Africa/El_Aaiun","Africa/Freetown","Africa/Gaborone","Africa/Harare","Africa/Johannesburg","Africa/Juba","Africa/Kampala","Africa/Khartoum","Africa/Kigali","Africa/Kinshasa","Africa/Lagos","Africa/Libreville","Africa/Lome","Africa/Luanda","Africa/Lubumbashi","Africa/Lusaka","Africa/Malabo","Africa/Maputo","Africa/Maseru","Africa/Mbabane","Africa/Mogadishu","Africa/Monrovia","Africa/Nairobi","Africa/Ndjamena","Africa/Niamey","Africa/Nouakchott","Africa/Ouagadougou","Africa/Porto-Novo","Africa/Sao_Tome","Africa/Tripoli","Africa/Tunis","Africa/Windhoek","America/Adak","America/Anchorage","America/Anguilla","America/Antigua","America/Araguaina","America/Argentina/Buenos_Aires","America/Argentina/Catamarca","America/Argentina/Cordoba","America/Argentina/Jujuy","America/Argentina/La_Rioja","America/Argentina/Mendoza","America/Argentina/Rio_Gallegos","America/Argentina/Salta","America/Argentina/San_Juan","America/Argentina/San_Luis","America/Argentina/Tucuman","America/Argentina/Ushuaia","America/Aruba","America/Asuncion","America/Atikokan","America/Bahia","America/Bahia_Banderas","America/Barbados","America/Belem","America/Belize","America/Blanc-Sablon","America/Boa_Vista","America/Bogota","America/Boise","America/Cambridge_Bay","America/Campo_Grande","America/Cancun","America/Caracas","America/Cayenne","America/Cayman","America/Chicago","America/Chihuahua","America/Costa_Rica","America/Creston","America/Cuiaba","America/Curacao","America/Danmarkshavn","America/Dawson","America/Dawson_Creek","America/Denver","America/Detroit","America/Dominica","America/Edmonton","America/Eirunepe","America/El_Salvador","America/Fort_Nelson","America/Fortaleza","America/Glace_Bay","America/Godthab","America/Goose_Bay","America/Grand_Turk","America/Grenada","America/Guadeloupe","America/Guatemala","America/Guayaquil","America/Guyana","America/Halifax","America/Havana","America/Hermosillo","America/Indiana/Indianapolis","America/Indiana/Knox","America/Indiana/Marengo","America/Indiana/Petersburg","America/Indiana/Tell_City","America/Indiana/Vevay","America/Indiana/Vincennes","America/Indiana/Winamac","America/Inuvik","America/Iqaluit","America/Jamaica","America/Juneau","America/Kentucky/Louisville","America/Kentucky/Monticello","America/Kralendijk","America/La_Paz","America/Lima","America/Los_Angeles","America/Lower_Princes","America/Maceio","America/Managua","America/Manaus","America/Marigot","America/Martinique","America/Matamoros","America/Mazatlan","America/Menominee","America/Merida","America/Metlakatla","America/Mexico_City","America/Miquelon","America/Moncton","America/Monterrey","America/Montevideo","America/Montserrat","America/Nassau","America/New_York","America/Nipigon","America/Nome","America/Noronha","America/North_Dakota/Beulah","America/North_Dakota/Center","America/North_Dakota/New_Salem","America/Ojinaga","America/Panama","America/Pangnirtung","America/Paramaribo","America/Phoenix","America/Port-au-Prince","America/Port_of_Spain","America/Porto_Velho","America/Puerto_Rico","America/Punta_Arenas","America/Rainy_River","America/Rankin_Inlet","America/Recife","America/Regina","America/Resolute","America/Rio_Branco","America/Santarem","America/Santiago","America/Santo_Domingo","America/Sao_Paulo","America/Scoresbysund","America/Sitka","America/St_Barthelemy","America/St_Johns","America/St_Kitts","America/St_Lucia","America/St_Thomas","America/St_Vincent","America/Swift_Current","America/Tegucigalpa","America/Thule","America/Thunder_Bay","America/Tijuana","America/Toronto","America/Tortola","America/Vancouver","America/Whitehorse","America/Winnipeg","America/Yakutat","America/Yellowknife","Antarctica/Casey","Antarctica/Davis","Antarctica/DumontDUrville","Antarctica/Macquarie","Antarctica/Mawson","Antarctica/McMurdo","Antarctica/Palmer","Antarctica/Rothera","Antarctica/Syowa","Antarctica/Troll","Antarctica/Vostok","Arctic/Longyearbyen","Asia/Aden","Asia/Almaty","Asia/Amman","Asia/Anadyr","Asia/Aqtau","Asia/Aqtobe","Asia/Ashgabat","Asia/Atyrau","Asia/Baghdad","Asia/Bahrain","Asia/Baku","Asia/Bangkok","Asia/Barnaul","Asia/Beirut","Asia/Bishkek","Asia/Brunei","Asia/Chita","Asia/Choibalsan","Asia/Colombo","Asia/Damascus","Asia/Dhaka","Asia/Dili","Asia/Dubai","Asia/Dushanbe","Asia/Famagusta","Asia/Gaza","Asia/Hebron","Asia/Ho_Chi_Minh","Asia/Hong_Kong","Asia/Hovd","Asia/Irkutsk","Asia/Jakarta","Asia/Jayapura","Asia/Jerusalem","Asia/Kabul","Asia/Kamchatka","Asia/Karachi","Asia/Kathmandu","Asia/Khandyga","Asia/Kolkata","Asia/Krasnoyarsk","Asia/Kuala_Lumpur","Asia/Kuching","Asia/Kuwait","Asia/Macau","Asia/Magadan","Asia/Makassar","Asia/Manila","Asia/Muscat","Asia/Nicosia","Asia/Novokuznetsk","Asia/Novosibirsk","Asia/Omsk","Asia/Oral","Asia/Phnom_Penh","Asia/Pontianak","Asia/Pyongyang","Asia/Qatar","Asia/Qostanay","Asia/Qyzylorda","Asia/Riyadh","Asia/Sakhalin","Asia/Samarkand","Asia/Seoul","Asia/Shanghai","Asia/Singapore","Asia/Srednekolymsk","Asia/Taipei","Asia/Tashkent","Asia/Tbilisi","Asia/Tehran","Asia/Thimphu","Asia/Tokyo","Asia/Tomsk","Asia/Ulaanbaatar","Asia/Urumqi","Asia/Ust-Nera","Asia/Vientiane","Asia/Vladivostok","Asia/Yakutsk","Asia/Yangon","Asia/Yekaterinburg","Asia/Yerevan","Atlantic/Azores","Atlantic/Bermuda","Atlantic/Canary","Atlantic/Cape_Verde","Atlantic/Faroe","Atlantic/Madeira","Atlantic/Reykjavik","Atlantic/South_Georgia","Atlantic/St_Helena","Atlantic/Stanley","Australia/Adelaide","Australia/Brisbane","Australia/Broken_Hill","Australia/Currie","Australia/Darwin","Australia/Eucla","Australia/Hobart","Australia/Lindeman","Australia/Lord_Howe","Australia/Melbourne","Australia/Perth","Australia/Sydney","Canada/Atlantic","Canada/Central","Canada/Eastern","Canada/Mountain","Canada/Newfoundland","Canada/Pacific","Europe/Amsterdam","Europe/Andorra","Europe/Astrakhan","Europe/Athens","Europe/Belgrade","Europe/Berlin","Europe/Bratislava","Europe/Brussels","Europe/Bucharest","Europe/Budapest","Europe/Busingen","Europe/Chisinau","Europe/Copenhagen","Europe/Dublin","Europe/Gibraltar","Europe/Guernsey","Europe/Helsinki","Europe/Isle_of_Man","Europe/Istanbul","Europe/Jersey","Europe/Kaliningrad","Europe/Kiev","Europe/Kirov","Europe/Lisbon","Europe/Ljubljana","Europe/London","Europe/Luxembourg","Europe/Madrid","Europe/Malta","Europe/Mariehamn","Europe/Minsk","Europe/Monaco","Europe/Moscow","Europe/Oslo","Europe/Paris","Europe/Podgorica","Europe/Prague","Europe/Riga","Europe/Rome","Europe/Samara","Europe/San_Marino","Europe/Sarajevo","Europe/Saratov","Europe/Simferopol","Europe/Skopje","Europe/Sofia","Europe/Stockholm","Europe/Tallinn","Europe/Tirane","Europe/Ulyanovsk","Europe/Uzhgorod","Europe/Vaduz","Europe/Vatican","Europe/Vienna","Europe/Vilnius","Europe/Volgograd","Europe/Warsaw","Europe/Zagreb","Europe/Zaporozhye","Europe/Zurich","GMT","Indian/Antananarivo","Indian/Chagos","Indian/Christmas","Indian/Cocos","Indian/Comoro","Indian/Kerguelen","Indian/Mahe","Indian/Maldives","Indian/Mauritius","Indian/Mayotte","Indian/Reunion","Pacific/Apia","Pacific/Auckland","Pacific/Bougainville","Pacific/Chatham","Pacific/Chuuk","Pacific/Easter","Pacific/Efate","Pacific/Enderbury","Pacific/Fakaofo","Pacific/Fiji","Pacific/Funafuti","Pacific/Galapagos","Pacific/Gambier","Pacific/Guadalcanal","Pacific/Guam","Pacific/Honolulu","Pacific/Kiritimati","Pacific/Kosrae","Pacific/Kwajalein","Pacific/Majuro","Pacific/Marquesas","Pacific/Midway","Pacific/Nauru","Pacific/Niue","Pacific/Norfolk","Pacific/Noumea","Pacific/Pago_Pago","Pacific/Palau","Pacific/Pitcairn","Pacific/Pohnpei","Pacific/Port_Moresby","Pacific/Rarotonga","Pacific/Saipan","Pacific/Tahiti","Pacific/Tarawa","Pacific/Tongatapu","Pacific/Wake","Pacific/Wallis","US/Alaska","US/Arizona","US/Central","US/Eastern","US/Hawaii","US/Mountain","US/Pacific","UTC"]
CheckSaveState = True #@param {type:"boolean"}
OptimizeColab = True #@param {type:"boolean"}
UseSameAccountAsMount = True #@param {type:"boolean"}

color = {
   "purple":'\033[95m',
   "cyan":'\033[96m',
   "darkcyan":'\033[36m',
   "blue":'\033[94m',
   "green":'\033[92m',
   "yellow":'\033[93m',
   "red":'\033[91m',
   "bold":'\033[1m',
   "underline":'\033[4m',
   "end":'\033[0m'
   }

print(f"{color['bold']}{color['underline']}Initializing Environment{color['end']}:")
with output.use_tags('DriveAuthentication'):
    print(f"\n*Drive Authentication: {color['yellow']}Requesting.{color['end']}" , end="")
    print('\n')

    if UseSameAccountAsMount:
        from google.colab import drive
        drive.mount("/content/drive")
    else:    
        !add-apt-repository -y ppa:alessandro-strada/ppa 2>&1 > /dev/null
        !apt-get update -qq 2>&1 > /dev/null
        !apt-get -y install -qq google-drive-ocamlfuse fuse
        from google.colab import auth
        auth.authenticate_user()
        from oauth2client.client import GoogleCredentials
        creds = GoogleCredentials.get_application_default()
        import getpass
        !google-drive-ocamlfuse -headless -id={creds.client_id} -secret={creds.client_secret} < /dev/null 2>&1 | grep URL
        vcode = getpass.getpass()
        !echo {vcode} | google-drive-ocamlfuse -headless -id={creds.client_id} -secret={creds.client_secret}
        %cd /content
        !mkdir drive
        %cd drive
        !mkdir MyDrive
        %cd ..
        %cd ..
        !google-drive-ocamlfuse /content/drive/MyDrive

output.clear(output_tags='DriveAuthentication')
print(f"\n*Drive Authentication:{color['green']} Done{color['end']}")

with output.use_tags('SearchEngine'):
    print(f"\n*1337x Search Engine: {color['yellow']}Preparing.{color['end']}" , end="")
    print('\n')

    !pip install 1337x

output.clear(output_tags='SearchEngine')
from py1337x import py1337x
print(f"*1337x Search Engine:{color['green']} Done{color['end']}")


with output.use_tags('torPackage'):
    print(f"\n*Torrent Client Package: {color['yellow']}Preparing{color['end']}", end="")
    print('\n')

    !pip install python-libtorrent-bin #&> /dev/null
    !pip install magneturi #&> /dev/null

output.clear(output_tags='torPackage')
import libtorrent as lt
import magneturi
print(f"*Torrent Client Package: {color['green']}Done{color['end']}")
    

# Server Start
ses = lt.session()
ses.listen_on(6881, 6891)

# Torrent States
state_str = [
    "Queued",
    "Checking",
    "Downloading metadata",
    "Downloading",
    "Finished",
    "Seeding",
    "Allocating",
    "Checking fastresume",
]

symbol = {
    "Queued" : "‚ßó",
    "Checking" : "‚ôª",
    "Downloading metadata" :"‚öù",
    "Downloading" : "ü°á",
    "Finished" : "‚àû",
    "Seeding" : "‚Üπ",
    "Allocating" : "÷é",
    "Checking fastresume" :"‚óå",
    
    "checkmark" : "‚úÖ",
    "star" : "‚úØ",
    "seeder" : "‚óï",
    "speed" : "‚áÖ",
    "file" : "‚éô",
    "warning": "‚ö†",
    "skull" :"üï±",
    "arrowUp" : "‚Ø≠",
    "arrowDown" : "‚ØØ",
    "time" : "‚òØ",
    "cross" : "‚®≥"
} 

seedQuality = {
    "low" : "‚ô¶Ô∏è",
    "med" : "üî∂",
    "high" : "üü¢"
}

animDownload = {
    1 : "ü°£",
    0 : "ü°≥",
    2 : "ü¢É"

}

animUpload = {
    2 : "ü°°",
    1 : "ü°±",
    0 : "ü¢Å"
}

searchCategory = {
    "none" : "",
    "movies" : "movies",
    "tv" : "tv",
    "games" : "games",
    "music" : "music",
    "apps" : "apps",
    "anime" : "anime",
    "documentaries" : "documentaries",
    "porn" : "xxx",
    "others" :"others",

}

searchSortBy = {
    "time" : "time",
    "size" : "size",
    "seeders" : "seeders",
    "leechers" : "leechers",
}

searchSortOrder = {
    "descending" : "desc",
    "ascending" : "asc"
}

torrents = py1337x(proxy='1337xx.to', cookie='<cookie>', cache='py1337xCache', cacheTime=86400, backend='memory') #change proxy if it doesnt work
downloads = []
finished = []
Logs = []

TorrentHistory = Downloads + '/TorrentHistory/'
completed_path = TorrentHistory + 'Completed/'
deleted_path = TorrentHistory + 'Deleted/'
params = {"save_path": SaveDirectory}

for path in [TorrentHistory, SaveDirectory, completed_path]:
 if not os.path.exists(path):
	  os.makedirs(path) 



#######################Functions##################

##########################UI######################
class UI:

    @staticmethod
    def bannerTorrentDownloader():
        print(f"")
        print(f"{color['red']}‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó{color['end']}‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë")
        print(f"{color['red']}‚ïö‚ïê‚ïê‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù{color['end']}‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïë‚ïö‚ïê‚ïê‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó")
        print(f"‚ñë‚ñë‚ñë{color['red']}‚ñà‚ñà‚ïë{color['end']}‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïî‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ïö‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïî‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù")
        print(f"‚ñë‚ñë‚ñë{color['red']}‚ñà‚ñà‚ïë{color['end']}‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ïî‚ïê‚ñà‚ñà‚ñà‚ñà‚ïë‚ñë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó")
        print(f"‚ñë‚ñë‚ñë{color['red']}‚ñà‚ñà‚ïë{color['end']}‚ñë‚ñë‚ñë‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ïö‚ñà‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñë‚ñë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñë‚ñà‚ñà‚ïë‚ñë‚ïö‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë")
        print(f"‚ñë‚ñë‚ñë{color['red']}‚ïö‚ïê‚ïù{color['end']}‚ñë‚ñë‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïù‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù")
    
    @staticmethod
    def bannerTorrentSearcher():
        
        print(f"‚ñë‚ñÄ‚ñà‚ñÄ‚ÄÉ‚ñà‚ñÄ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñÄ‚ÄÉ‚ñà‚ñÑ‚ñë‚ñà‚ÄÉ‚ñÄ‚ñà‚ñÄ‚ÄÉ ‚ÄÉ‚ñà‚ñÄ‚ÄÉ‚ñà‚ñÄ‚ñÄ‚ÄÉ‚ñÑ‚ñÄ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñÄ‚ÄÉ‚ñà‚ñë‚ñà")
        print(f"‚ñë‚ñë‚ñà‚ñë‚ÄÉ‚ñà‚ñÑ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñÑ‚ÄÉ‚ñà‚ñÄ‚ñÑ‚ÄÉ‚ñà‚ñà‚ñÑ‚ÄÉ‚ñà‚ñë‚ñÄ‚ñà‚ÄÉ‚ñë‚ñà‚ñë‚ÄÉ ‚ÄÉ‚ñÑ‚ñà‚ÄÉ‚ñà‚ñà‚ñÑ‚ÄÉ‚ñà‚ñÄ‚ñà‚ÄÉ‚ñà‚ñÄ‚ñÑ‚ÄÉ‚ñà‚ñÑ‚ñÑ‚ÄÉ‚ñà‚ñÄ‚ñà")
    
    @staticmethod
    def display_msg(messageType,message):
        msgType = ""
        colorType = ""
        if messageType.lower() == "error":
            msgType = "Error"
            colorType = color['red']
        elif messageType.lower() == "success":
            msgType = "Success"
            colorType = color['green']
        elif messageType.lower() == "warning":
            msgType = "Warning"
            colorType = color['yellow']
        elif messageType.lower() == "check":
            msgType = "Checking"
            colorType = color['yellow']
        else:
            msgType = "Missing MsgType"
            colorType = color['purple']
        
        return f"{colorType}{msgType}{color['end']}:{message}"


    @staticmethod
    def set_userInput(messageConsole,message):
        print(f"{color['yellow']}{message}{color['darkcyan']}")
        userInput = input(f" {messageConsole}")
        print(f"{color['end']}")
        return userInput


    @staticmethod
    def display_items(Items): #Makeshift
        if Items:
            for index,data in enumerate(Items[:]):
                print(f" [{index+1}] {data}")
        else:
            print(f"{color['end']}{color['darkcyan']}Currently Empty....{color['end']}")

    @staticmethod
    def set_verifyInput(messageConsole,message,choices):
        isRunning = True
        with output.use_tags('verifyInput'):
            while isRunning:
                userInput = UI.set_userInput(messageConsole,message).lower()
                if userInput in choices:
                    isRunning = False
                else:
                    UI.display_msg("error","Invalid choice!")

        output.clear(output_tags='verifyInput')
        return userInput


    @staticmethod
    def set_verifyInputInt(messageConsole,message,min,max):
        isRunning = True
        with output.use_tags('verifyInt'):
            while isRunning:
                userInput = UI.set_userInput(messageConsole,message)
                if not userInput.isnumeric():
                    UI.display_msg("error",f"The input must be an integer!")
                    time.sleep(2)
                    continue
                
                userInput = int(userInput)

                if userInput >= min and userInput <= max :
                    isRunning = False
                
                UI.display_msg("error",f"It must be between {min} - {max}")
                time.sleep(2)    

        output.clear(output_tags='verifyInt')
        return userInput

    @staticmethod
    def set_verifyInputBool(message):
        isRunning = True
        choice = True
        with output.use_tags('verifyBool'):
            userInput = UI.set_userInput(f"Yes/No: ",f"{color['yellow']}{message}{color['end']}")
            while isRunning:
                userInput = userInput.lower()
                if userInput in ['y','n','yes','no']:
                    isRunning = False
                    if userInput in ['n','no']:
                        choice = False
                else:
                    userInput = UI.set_userInput(f"Yes/No",f"{color['red']}Error{color['end']}:{color['yellow']}Invalid Input, Please choose only Y/N{color['end']}")
            
        output.clear(output_tags='verifyBool')
        return choice
  
    @staticmethod
    def set_verifyInputAdv(messageConsole,message,choices,ifHasIndex = False,ifsetOfIntegers = False,min=0,max=0):
        filter = choices
        if ifsetOfIntegers:
            isRunning = True
            while isRunning:
                userInput = UI.set_userInput(messageConsole,message)
                if not userInput.isnumeric():
                    UI.display_msg("error",f"The input must be an integer!")
                    time.sleep(2)
                    continue
                if userInput >= min and userInput <= max :
                    isRunning = False
                print(f"{color['red']}Error{color['end']}: The input must be between {min}-{max}")
                UI.display_msg("error",f"It must be between {min} - {max}")
                time.sleep(2)
        else:
            userInput = UI.set_userInput(messageConsole,message)
            if ifHasIndex and userInput.isnumeric():
                filter = filter + GeneralFunc.generate_enumArray(len(choices),False)
                userInput = choices[int(userInput)-1]
            else:
                userInput = userInput.lower()
            while userInput not in filter:
                userInput = UI.set_userInput(messageConsole,message)
                
                if ifHasIndex and userInput.isnumeric():
                    userInput = choices[int(userInput)-1]
                else:
                    userInput = userInput.lower()

        return userInput

    @staticmethod
    def colorformatStart(index):
        print(f"{color['end'] if index % 2 == 0 else color['darkcyan']}",end="")
    
    @staticmethod
    def colorformatEnd():
        print(color['end'],end="")

    @staticmethod
    def display_chainBar():
        print(f"{color['bold']}{color['underline']}-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-{color['end']}")


#$$$$$$$$$$$$$GeneralFunc##############

class GeneralFunc:
    @staticmethod
    def generate_enumArray(max,ifStartWithZero = True):
        container = []
        shift = 1
        if not ifStartWithZero:
            shift = 0

        while max:
            container.append(max - shift)
            max -= 1
        return container

    @staticmethod
    def addExitToDict(items):
        array = []
        for data in items:
            array.append(data)
        array.append("exit")
        return array

    @staticmethod
    def timeCheck():
        country_time = datetime.now(pytz.timezone(Timezone))
        return country_time.strftime("%I:%M %p")

    @staticmethod
    def removeFromList(Items,data):
        Items.remove(data)

    @staticmethod
    def addFromList(Items,data):
        Items.append(data)
        
    @staticmethod
    def removeFromDisk(location):
        os.remove(location)
    
    @staticmethod
    def checkFromDisk(location):
        ifExist = False
        if os.path.exists(location):
            ifExist = True
        return ifExist

    @staticmethod
    def saveFromDisk(location,data):
         with open(location, 'wb') as fp:
            pickle.dump(data, fp)
    
    @staticmethod
    def readFromDisk(location):
        try:
            with open (location, 'rb') as fp:
                data = pickle.load(fp)
        except FileNotFoundError:
            print("File not found")
            return False

        return data

    @staticmethod
    def get_rate(val):
        prefix = ['B', 'kB', 'MB', 'GB', 'TB']
        for i in range(len(prefix)):
            if abs(val) < 1000:
                if i == 0:
                    return '%5.3g %s' % (val, prefix[i])
                else:
                    return '%4.3g %s' % (val, prefix[i])
            val /= 1000

        return '%6.3g PB' % val

    @staticmethod
    def get_eta(fileSize,speed):
        try:
            eta = fileSize / speed

        except ZeroDivisionError:
            return 0
            
        return eta 

    @staticmethod
    def get_secToTime(seconds):
        if seconds == 0:
            return f"."
        seconds = int(seconds)
        hh = str(seconds//3600)
        mm = str((seconds%3600)//60)
        ss = str((seconds%3600)%60)
        message = ""
        
        if int(hh) > 24:
            message = f"{int(int(hh)/ 24)} days."
        
        elif int(hh) :
            message = f"{hh}:{'0' if int(mm) < 10 else ''}{mm}:{ss}"
        else:
            message = f"{'' if int(hh) == 0 else hh+ ' hours '}{'' if int(mm) == 0 else mm + 'm '}{'' if int(ss) == 0 else ss + 's.'}"
        
        return message

    @staticmethod
    def ifDuplicateFound(data):
        found = False
        for index, download in enumerate(downloads[:]):
            if download.name() == data:
                found = True
                print(f" {color['red']}Error{color['end']}: Torrent Already Existed on download Index: {index +1}. File: {color['underline']}{download.name()[:40]}...{color['end']}")
                break
        for item in finished:
            if item == data:
                found = True
                print(f" {color['red']}Error{color['end']}: Torrent Already Downloaded. File: {color['underline']}{data[:50]}...{color['end']}")
                break

        return found

    @staticmethod
    def transferTorrent(torrentName,transferType = "complete"):
        if transferType == "complete":
            try:
              shutil.move(TorrentHistory + torrentName + '.torrent', completed_path + torrentName + '.torrent')
            except:
              pass
        elif transferType == "delete":
            try:
              shutil.move(TorrentHistory + torrentName, deleted_path + torrentName)
            except:
              UI.display_msg("error",f"{torrentName} {color['yellow']} not found{color['end']}")
        else:
            UI.display_msg("error","Unknown Location")
          




##########################Main Functions########################################################
class MainFunc:
   
    @staticmethod
    def set_addTorrent(link):
        params = {"save_path": SaveDirectory}
        downloads.append(
            lt.add_magnet_uri(ses, link, params)
        )
    
    @staticmethod
    def set_verifyTorrentImport(magnetLink):
        if MainFunc.process_magnet(magnetLink):
            MainFunc.set_addTorrent(magnetLink)
    
    @staticmethod
    def set_inputMagnet():
        isRunning = True
        magnetLink = ""
        while isRunning:
            clear_output()
            UI_DownloadMenu.display_downloadQueue()
            magnetLink = UI.set_userInput("Input: ", "Enter Magnet Link Or Type Exit: ")
            if magnetLink.lower() == "exit":
                isRunning = False
            MainFunc.set_verifyTorrentImport(magnetLink)
            del magnetLink

    
    @staticmethod
    def set_importTorrent():
        try:
            while True:
                clear_output() 
                UI_DownloadMenu.display_downloadQueue()
                source = files.upload()
                filename = next(iter(source))
                if MainFunc.process_torrent(filename) == "NULL":
                    del source
                    break

                magnetLink = MainFunc.process_torrent(filename)
                if MainFunc.process_magnet(magnetLink):
                    downloads.append(lt.add_magnet_uri(ses, magnetLink, params))
                del source
        except StopIteration:
            pass
            

    @staticmethod
    def remove_torrent(download):
        ses.remove_torrent(download)
        downloads.remove(download)


    @staticmethod
    def remove_torrentDownload():
        clear_output()
        i = 0
        if not downloads:
            UI.display_msg("error",f"Download Queue is {color['red']}Empty{color['end']}, there's nothing to remove.")
        else:
            print(f"{color['yellow']}Download Queue:{color['end']}{color['end']}{color['darkcyan']} ")
            UI_DownloadMenu.display_downloadQueue()
            print(f"{color['end']}")
            try:
                choice = UI.set_verifyInputInt("Index: ","Choose what index to remove:",1,len(downloads))
                choice -= 1
                
            except:
                UI.display_msg("error","Invalid input!")

            for index, download in enumerate(downloads[:]):
                if choice == index:
                    MainFunc.remove_torrent(download)
                    UI.display_msg("success",f"{color['underline']}{download.name()}{color['cyan']}Removed{color['end']}")
                    GeneralFunc.addFromList(Logs,f"   {color['red']}{symbol['cross']}{color['end']}{color['red']} Removed{color['end']}: {download.name()} is {color['cyan']}removed{color['end']} on {color['cyan']}{GeneralFunc.timeCheck()}{color['end']}")
                    break
            else :
                UI.display_msg("error",f"Torrent index[{choice}] is {color['red']}not found{color['end']}")
        
                        

    @staticmethod
    def process_torrent(file):
        try:
            magnet = magneturi.from_torrent_file(file)
        except SyntaxError:
            print(f"{color['red']}Error: {color['end']}{file} is not a valid bittorent file. ")
            magnet = "NULL"
            return magnet
        
        except AssertionError:
            if os.path.exists(file):
                print(f"{color['yellow']}Error:{color['end']}{file} is already Completed therefore not added to queue. ")
            else:
                print(f"{color['red']}Error: {color['end']}{file} not Found.")
            magnet = "NULL"
            return magnet

        except : #Most Commonly DecodingException
            print(f"{color['red']}Error: {color['end']}Torrent is currupted, please download it again. Possible causes {color['yellow']}google drive quota exceeded.{color['end']} Filename: {file}")
            magnet = "NULL"
        
        return magnet


    @staticmethod
    def process_magnet(magnetLink,ifSaveToHistory = True):
        try:
            params = lt.parse_magnet_uri(magnetLink)
        except:
            print(f"\n{color['red']}Error{color['end']}: Only valid Magnet link is accepted!")
            time.sleep(3)
            return False
        session = lt.session()
        handle = session.add_torrent(params)
        timer = 0
        print(f"{color['green']}Downloading Metadata{color['end']}", end='')
        while not handle.has_metadata(): 
            timer = timer + 1
            time.sleep(0.1)
            if timer == 100:
                print(".",end='')
            elif timer == 250:
                print(".",end='')
            elif timer == 500:
                print(".",end='')
            elif timer == 750:
                print(".",end='')
            elif timer == 850:
                print(f"It\'s slow due to {color['red']}Low Seeders{color['end']}")

            if timer >= 900: #1 min estimation
                print(f"\n{color['red']}Timeout Error{color['end']}, can\'t download torrent metadata. Try again next time.")
                time.sleep(3)
                return False 
        torrent_info = handle.get_torrent_info()
        if GeneralFunc.ifDuplicateFound(torrent_info.name()):
            time.sleep(3)
            return False

        if ifSaveToHistory:
            torrent_file = lt.create_torrent(torrent_info)
            torrent_path = os.path.join(TorrentHistory, torrent_info.name() + ".torrent")
          
            with open(torrent_path, "wb") as f:
                f.write(lt.bencode(torrent_file.generate()))
            print (f"\n {torrent_info.name()} is {color['green']}Added")
            time.sleep(1)
        return True
    
    @staticmethod
    def set_addFileToQueue(fileName,ifCheckMetadata):
        file = TorrentHistory + fileName
        magnetLink = MainFunc.process_torrent(file)
        if magnetLink == "NULL":
            return False
        if ifCheckMetadata:
            UI.display_msg("check",f"Checking: {fileName} ")
            ifFound = MainFunc.process_magnet(magnetLink,ifSaveToHistory = False)
            if ifFound:
                downloads.append(lt.add_magnet_uri(ses, magnetLink, params))
                UI.display_msg("success",f"Added to download queue: {color['underline']}{fileName[:40]}...{color['end']}")
                return ifFound
        else:
            fileName = fileName.split(".torrent")
            if not GeneralFunc.ifDuplicateFound(fileName[0]):
                downloads.append(lt.add_magnet_uri(ses, magnetLink, params))
                UI.display_msg("success",f"Added to queue: {fileName[0]}")
                return True
            else:
                return False

    @staticmethod
    def recentSaveUpdater():
        downloadQueue = []
        saveLocation = TorrentHistory + 'recents.dat'
        
        for index, download in enumerate(downloads[:]):
            GeneralFunc.addFromList(downloadQueue,download.name())

        if not downloadQueue:
            if GeneralFunc.checkFromDisk(saveLocation):
                GeneralFunc.removeFromDisk(saveLocation)
        
        else:
            GeneralFunc.saveFromDisk(saveLocation,downloadQueue)



    @staticmethod
    def recentSaveLoader():
        saveLocation = TorrentHistory + 'recents.dat'
       
        if not downloads:
            downloadQueue = []
            print(f"{color['yellow']}Searching for Previous Unfinished queue:")
            if GeneralFunc.checkFromDisk(saveLocation):
                downloadQueue = GeneralFunc.readFromDisk(saveLocation)
            
            else:
                print(f"{color['yellow']}No recent savedata found.{color['end']}")
                return  
            
            userInput = ""
            with output.use_tags('getRecentUnfinished'):
                if len(downloadQueue) > 0:
                    print("A previous download list is recovered:")
                    UI.display_items(downloadQueue)
                    userInput = UI.set_verifyInputBool("Do you wanna resume downloading these?")
                if userInput:
                    for data in downloadQueue:
                        MainFunc.set_addFileToQueue(f"{data}.torrent",False)
                else:
                    print(f"Found unfinished download queue {color['cyan']}discarded{color['end']}.")
            time.sleep(2)
            output.clear(output_tags='getRecentUnfinished')

#################@UI Download Menu#################

class UI_DownloadMenu:
    
    @staticmethod
    def process_choice(choice):
        if choice == 0:
            UI.display_msg("success","Program Exiting.")     
        elif choice == 1:
            SearchEngine = TorrentSearch()
            SearchEngine.main()               
        elif choice == 2:
            MainFunc.set_inputMagnet()
        elif choice == 3:
            MainFunc.set_importTorrent()
        elif choice == 4:
            x = File_Manager()
            x.main()
        else:
            UI.display_msg("error","Invalid Input, Choice = 0-4 ")
        return True
        
        
    @staticmethod
    def display_downloadQueue():
        if downloads:
            print (f"\n{color['green']}Download Queue{color['end']}:")
            for index, download in enumerate(downloads[:]):
                print (f"[{'0' if index < 10 else ''}{index + 1}] {download.name()}, Seeders:{color['green'] if download.status().num_seeds > 100 else color['yellow'] if download.status().num_seeds > 30 else color['red']}{download.status().num_seeds}{color['end']}{color['darkcyan']}(Not Realtime){color['end']}")
        else:
            print(f"{color['yellow']}Download Queue is currently {color['darkcyan']}Empty{color['end']}!")

    @staticmethod
    def display_logs():
        
        if Logs or downloads:
            print(f"{color['yellow']}\nRecent Logs{color['end']}:")
            if Logs:
                for data in Logs:
                    print(f"{data}")
        elif not downloads and not Logs:
            print(f"{color['yellow']}\nRecent Logs{color['end']}:{color['bold']} Currently Empty...{color['end']}")
            

    @staticmethod
    def display_menu():
        UI.display_chainBar()
        print(f"Menu:")
        print(f" [{color['bold']}{color['underline']}1{color['end']}]{color['cyan']}{color['underline']}Search Torrents{color['end']} | [{color['bold']}{color['underline']}2{color['end']}]{color['cyan']}{color['underline']}Add via Magnet{color['end']} | [{color['bold']}{color['underline']}3{color['end']}] {color['cyan']}{color['underline']}Add via .torrent{color['end']} | [{color['bold']}{color['underline']}4{color['end']}]{color['cyan']}{color['underline']}Manage Unfinished{color['end']} | [{color['bold']}{color['underline']}0{color['end']}]{color['cyan']}{color['underline']}Exit{color['end']}\n")
        #print(f"-=-=-=-=-=-=-=-=-=-=-=-=-=-=--=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-")
    
    @staticmethod
    def main():
        isRunning = True
        while isRunning:
            UI_DownloadMenu.display_menu()
            choice = UI.set_verifyInputInt("Choice: ","Enter Choice: ",0,4)
            if UI_DownloadMenu.process_choice(choice):
                isRunning = False

    def header():
        print(f"{color['bold']}{color['yellow']}Download Queue:{color['end']}")
        if not downloads:
            if not finished:
                print(f"{color['bold']}No Torrent Found, Please add one in the menu.\n{color['end']}")

    def set_inputMenu():
        choice = ""
        with output.use_tags('menu'):
            choice = input("[Default]Menu \t [R]Remove Torrent \t [Q]Quit\nEnter Magnet\\Choice: ")
        output.clear(output_tags='menu')
        print("\n")
        return choice

    def ifMagnet(data):
        isMagnet = False 
        if len(data) >= 11:
            if data[:11] == "magnet:?xt=":
                isMagnet = True
        return isMagnet

    def ifSearch(data):
        isSearch =False
        if len(data) >= 7:
            if str(data[:7]).lower() == "/search":
                isSearch = True
        return isSearch

    def processIfSearch(data):
        UI.display_chainBar()
        SearchEngine = TorrentSearch()
        SearchEngine.main(data[7:])

    
    def processIfMagnet(data):
        UI.display_chainBar()
        print(f"{color['yellow']}Processing Magnet: {color['darkcyan']}{color['underline']}{data[:8]}...{data[64:]}{color['end']}")
        MainFunc.set_verifyTorrentImport(data)

            
########################File Manager#####################
class File_Manager:

    def __init__(self):
      self._currentQueue = self.get_Unfinished()
      self._selectedDownloadQueue = []
      self._running = True

    def kill(self):
        self._running = False

    def add_toQueue(self,items,data):
        GeneralFunc.addFromList(items,data)

    def remove_toQueue(self,items,data):
        GeneralFunc.removeFromList(items,data)

    def delete_torrentFile(self, data):
        print(f"{color['yellow']}Removed: {color['darkcyan']}{data}{color['end']}")
        GeneralFunc.transferTorrent(data,"delete")

    def get_Unfinished(self):
        paths = os.listdir(TorrentHistory)
        return [path for path in paths if path.endswith('.torrent')]

    def displayQueue(self,items):
        UI.display_items(items)
    
    def selectAllList(self):
        self._selectedDownloadQueue += self._currentQueue
        self._currentQueue = []

    def restoreDefaultList(self):
        self._currentQueue = self.get_Unfinished()
        self._selectedDownloadQueue = []
      
    def updateQueue(self,data, updateType):
        if updateType == "remove":
            self.add_toQueue(self._currentQueue,data)
            self.remove_toQueue(self._selectedDownloadQueue,data)
        elif updateType == "add":
            self.add_toQueue(self._selectedDownloadQueue,data)
            self.remove_toQueue(self._currentQueue,data)
        else:
            UI.display_msg("error","Unknown updateType") #for redundancy

          
    def clear_unfinishedTorrent(self):
        if not self._currentQueue:
           UI.display_msg("warning","No recent downloads found.")
        else:
            choice = UI.set_verifyInputBool("Are you sure you want to clear recents? ")
            print(f"{color['yellow']}These files will be moved in Deleted folder{color['end']}")
            if choice:
                for data in self._currentQueue:
                    self.delete_torrentFile(data) 
            time.sleep(3) 

    def menu_removeFromQueue(self):
        clear_output()
        if self._selectedDownloadQueue:
            print(f"{color['yellow']}Selected Items:{color['end']}{color['cyan']}")
            self.displayQueue(self._selectedDownloadQueue)
            print(color['end'])
            choice = UI.set_verifyInputInt("Choice: ","Input index to remove: ",1,len(self._selectedDownloadQueue))
            self.updateQueue(self._selectedDownloadQueue[choice - 1],"remove")
        else:
            print(f"There's nothing to remove...")
            time.sleep(2)

    def finalize(self):
        if len(self._selectedDownloadQueue):
            print(f"{color['yellow']}Do you want to verify {'these torrents' if len(self._selectedDownloadQueue) > 1 else 'this torrent'} for seeders?{color['darkcyan']}(May take longer){color['end']}")
            checkMetadata = UI.set_verifyInputBool("Check for Verification?")
            invalid = 0
            for data in self._selectedDownloadQueue:
                print(f"{color['yellow']}File:{color['end']} {color['underline']}{color['cyan']}{data}{color['end']}")
                if not MainFunc.set_addFileToQueue(data,checkMetadata):
                    invalid = invalid + 1
                else:
                    print(f"{color['green']} Added:{color['end']} {color['underline']}{color['cyan']}{data}{color['end']}")
            print(f"{color['yellow']}Import complete. Errors found: {invalid} {color['end']}")
        else:
            print(f"No data is imported")
                      
    def main(self):
        clear_output()
        if len(self._currentQueue): 
            while self._running:
                clear_output()
                print(f"{color['yellow']}Enter the Index to be added to the queue.{color['end']}")
                print(f"{color['yellow']}Found Unfinished Items:{color['end']}{color['darkcyan']}")
                self.displayQueue(self._currentQueue)
                print(f"{color['yellow']}Selected Items:{color['end']}{color['cyan']}")
                self.displayQueue(self._selectedDownloadQueue)
                print(f"{color['end']}\nChoice: [{color['underline']}A{color['end']}]Select All [{color['underline']}R{color['end']}]Remove from Queue [{color['underline']}D{color['end']}]Return Default List [{color['underline']}C{color['end']}]Clear all Unfinished [{color['underline']}X{color['end']}]Exit")
                print(f"{color['yellow']}Choice: Enter Choice:{color['end']}")
                userInput = input("Choose: ")
                
                if userInput.isnumeric():
                    userInput = int(userInput) - 1
                    if userInput >= 0 and userInput < len(self._currentQueue):
                        self.updateQueue(self._currentQueue[userInput],"add")
                    else:
                        print(f"{color['red']}Error:{color['end']}input must not exceed 0-{len(self._currentQueue)}")
                        time.sleep(2)
                elif userInput.lower() == 'r':
                    self.menu_removeFromQueue()
                elif userInput.lower() == 'd':
                    self.restoreDefaultList()
                elif userInput.lower() == 'c':
                    self.clear_unfinishedTorrent()
                elif userInput.lower() == 'a':
                    self.selectAllList()
                elif userInput.lower() == 'x':
                    self.finalize()
                    time.sleep(3)
                    self.kill()

        else:
            print("There's currently no unfinished download at the moment. Now go get some..")
            time.sleep(2)


######################@title Start Program################

class UI_Download:
    def __init__(self):
        self._running = True
        self._animationFrames = 0
        self._nextShift = 0
    
    # Thread Killing
    def kill(self):
        self._running = False
    def defaultNextShift(self):
        self._nextShift = 0
    def updateNextShift(self):
        self._nextShift += -1

    def get_nextShift(self):
        return self._nextShift
    
    def updateAnimFrames(self):
        if self._animationFrames >= 2:
            self._animationFrames = 0
        else: 
            self._animationFrames += 1

    def get_animationFrames(self):
        return self._animationFrames

    # Print Status Bar
    def completeDownload(self,name):
        currentMessage = f"   {color['yellow']}{symbol['star']}{color['end']}{color['green']} Completed{color['end']}: {name} {color['cyan']}completed{color['end']} on {color['cyan']}{GeneralFunc.timeCheck()}{color['end']}"
        print(currentMessage)
        GeneralFunc.addFromList(Logs,currentMessage)
        GeneralFunc.addFromList(finished,str(name))
        GeneralFunc.transferTorrent(name,"complete")



    def show(self) :
        MainFunc.recentSaveUpdater()
        layout = widgets.Layout(width="auto")
        style = {"description_width": "initial"}
        download_bars = [
            widgets.FloatSlider(
                step=0.01, disabled=True, layout=layout, style=style
            )
            for _ in downloads
        ]
        display(*download_bars)

        while self._running and downloads:
            self.defaultNextShift()
            for index, download in enumerate(downloads[:]):
                bar = download_bars[index + self.get_nextShift()]
                status = download.status()
                if not download.is_seed():
                    bar.value = status.progress * 100
                    bar.description = " ".join(
                        [
                            f"\t",
                            f"{symbol[state_str[status.state]]}",
                            f"{download.name()[:28]}...",
                            f" | {seedQuality['high'] if status.num_seeds >= 100 else seedQuality['med'] if status.num_seeds >=30 else seedQuality['low'] }Seeder:{status.num_seeds}",
                            f" | {animDownload[self.get_animationFrames()]} {GeneralFunc.get_rate(status.download_rate)}/s",
                            f" | {animUpload[self.get_animationFrames()]} {GeneralFunc.get_rate(status.upload_rate)}/s",
                            f" | Size:{GeneralFunc.get_rate(status.total_wanted_done)}/{GeneralFunc.get_rate(status.total_wanted)}",
                            f"{'| '+ symbol['time'] if state_str[status.state] == 'Downloading' else ''} {GeneralFunc.get_secToTime(GeneralFunc.get_eta(status.total_wanted - status.total_wanted_done,status.download_rate)) if state_str[status.state] == 'Downloading' else '.'}",
   
                        ]
                    )
                
                else:
                    self.completeDownload(download.name())
                    self.updateNextShift()
                    MainFunc.remove_torrent(download)
                    MainFunc.recentSaveUpdater()
                    bar.close()
                    download_bars.remove(bar)
                
                self.updateAnimFrames()
            
            time.sleep(.5)
            
        
        if finished and not downloads:
            print(f"\n{symbol['checkmark']} {color['green']}All Pending Downloads are Completed, Consider adding some!{color['end']}")

```
