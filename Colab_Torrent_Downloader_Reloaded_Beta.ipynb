```python
#@title â†Initialize Downloader{ run: "auto" }
DEBUG = False
from google.colab import output
Downloads = "/content/drive/MyDrive/Downloads" #@param {type:"string"}
Timezone = 'Asia/Manila' #@param ["Africa/Abidjan","Africa/Accra","Africa/Addis_Ababa","Africa/Algiers","Africa/Asmara","Africa/Bamako","Africa/Bangui","Africa/Banjul","Africa/Bissau","Africa/Blantyre","Africa/Brazzaville","Africa/Bujumbura","Africa/Cairo","Africa/Casablanca","Africa/Ceuta","Africa/Conakry","Africa/Dakar","Africa/Dar_es_Salaam","Africa/Djibouti","Africa/Douala","Africa/El_Aaiun","Africa/Freetown","Africa/Gaborone","Africa/Harare","Africa/Johannesburg","Africa/Juba","Africa/Kampala","Africa/Khartoum","Africa/Kigali","Africa/Kinshasa","Africa/Lagos","Africa/Libreville","Africa/Lome","Africa/Luanda","Africa/Lubumbashi","Africa/Lusaka","Africa/Malabo","Africa/Maputo","Africa/Maseru","Africa/Mbabane","Africa/Mogadishu","Africa/Monrovia","Africa/Nairobi","Africa/Ndjamena","Africa/Niamey","Africa/Nouakchott","Africa/Ouagadougou","Africa/Porto-Novo","Africa/Sao_Tome","Africa/Tripoli","Africa/Tunis","Africa/Windhoek","America/Adak","America/Anchorage","America/Anguilla","America/Antigua","America/Araguaina","America/Argentina/Buenos_Aires","America/Argentina/Catamarca","America/Argentina/Cordoba","America/Argentina/Jujuy","America/Argentina/La_Rioja","America/Argentina/Mendoza","America/Argentina/Rio_Gallegos","America/Argentina/Salta","America/Argentina/San_Juan","America/Argentina/San_Luis","America/Argentina/Tucuman","America/Argentina/Ushuaia","America/Aruba","America/Asuncion","America/Atikokan","America/Bahia","America/Bahia_Banderas","America/Barbados","America/Belem","America/Belize","America/Blanc-Sablon","America/Boa_Vista","America/Bogota","America/Boise","America/Cambridge_Bay","America/Campo_Grande","America/Cancun","America/Caracas","America/Cayenne","America/Cayman","America/Chicago","America/Chihuahua","America/Costa_Rica","America/Creston","America/Cuiaba","America/Curacao","America/Danmarkshavn","America/Dawson","America/Dawson_Creek","America/Denver","America/Detroit","America/Dominica","America/Edmonton","America/Eirunepe","America/El_Salvador","America/Fort_Nelson","America/Fortaleza","America/Glace_Bay","America/Godthab","America/Goose_Bay","America/Grand_Turk","America/Grenada","America/Guadeloupe","America/Guatemala","America/Guayaquil","America/Guyana","America/Halifax","America/Havana","America/Hermosillo","America/Indiana/Indianapolis","America/Indiana/Knox","America/Indiana/Marengo","America/Indiana/Petersburg","America/Indiana/Tell_City","America/Indiana/Vevay","America/Indiana/Vincennes","America/Indiana/Winamac","America/Inuvik","America/Iqaluit","America/Jamaica","America/Juneau","America/Kentucky/Louisville","America/Kentucky/Monticello","America/Kralendijk","America/La_Paz","America/Lima","America/Los_Angeles","America/Lower_Princes","America/Maceio","America/Managua","America/Manaus","America/Marigot","America/Martinique","America/Matamoros","America/Mazatlan","America/Menominee","America/Merida","America/Metlakatla","America/Mexico_City","America/Miquelon","America/Moncton","America/Monterrey","America/Montevideo","America/Montserrat","America/Nassau","America/New_York","America/Nipigon","America/Nome","America/Noronha","America/North_Dakota/Beulah","America/North_Dakota/Center","America/North_Dakota/New_Salem","America/Ojinaga","America/Panama","America/Pangnirtung","America/Paramaribo","America/Phoenix","America/Port-au-Prince","America/Port_of_Spain","America/Porto_Velho","America/Puerto_Rico","America/Punta_Arenas","America/Rainy_River","America/Rankin_Inlet","America/Recife","America/Regina","America/Resolute","America/Rio_Branco","America/Santarem","America/Santiago","America/Santo_Domingo","America/Sao_Paulo","America/Scoresbysund","America/Sitka","America/St_Barthelemy","America/St_Johns","America/St_Kitts","America/St_Lucia","America/St_Thomas","America/St_Vincent","America/Swift_Current","America/Tegucigalpa","America/Thule","America/Thunder_Bay","America/Tijuana","America/Toronto","America/Tortola","America/Vancouver","America/Whitehorse","America/Winnipeg","America/Yakutat","America/Yellowknife","Antarctica/Casey","Antarctica/Davis","Antarctica/DumontDUrville","Antarctica/Macquarie","Antarctica/Mawson","Antarctica/McMurdo","Antarctica/Palmer","Antarctica/Rothera","Antarctica/Syowa","Antarctica/Troll","Antarctica/Vostok","Arctic/Longyearbyen","Asia/Aden","Asia/Almaty","Asia/Amman","Asia/Anadyr","Asia/Aqtau","Asia/Aqtobe","Asia/Ashgabat","Asia/Atyrau","Asia/Baghdad","Asia/Bahrain","Asia/Baku","Asia/Bangkok","Asia/Barnaul","Asia/Beirut","Asia/Bishkek","Asia/Brunei","Asia/Chita","Asia/Choibalsan","Asia/Colombo","Asia/Damascus","Asia/Dhaka","Asia/Dili","Asia/Dubai","Asia/Dushanbe","Asia/Famagusta","Asia/Gaza","Asia/Hebron","Asia/Ho_Chi_Minh","Asia/Hong_Kong","Asia/Hovd","Asia/Irkutsk","Asia/Jakarta","Asia/Jayapura","Asia/Jerusalem","Asia/Kabul","Asia/Kamchatka","Asia/Karachi","Asia/Kathmandu","Asia/Khandyga","Asia/Kolkata","Asia/Krasnoyarsk","Asia/Kuala_Lumpur","Asia/Kuching","Asia/Kuwait","Asia/Macau","Asia/Magadan","Asia/Makassar","Asia/Manila","Asia/Muscat","Asia/Nicosia","Asia/Novokuznetsk","Asia/Novosibirsk","Asia/Omsk","Asia/Oral","Asia/Phnom_Penh","Asia/Pontianak","Asia/Pyongyang","Asia/Qatar","Asia/Qostanay","Asia/Qyzylorda","Asia/Riyadh","Asia/Sakhalin","Asia/Samarkand","Asia/Seoul","Asia/Shanghai","Asia/Singapore","Asia/Srednekolymsk","Asia/Taipei","Asia/Tashkent","Asia/Tbilisi","Asia/Tehran","Asia/Thimphu","Asia/Tokyo","Asia/Tomsk","Asia/Ulaanbaatar","Asia/Urumqi","Asia/Ust-Nera","Asia/Vientiane","Asia/Vladivostok","Asia/Yakutsk","Asia/Yangon","Asia/Yekaterinburg","Asia/Yerevan","Atlantic/Azores","Atlantic/Bermuda","Atlantic/Canary","Atlantic/Cape_Verde","Atlantic/Faroe","Atlantic/Madeira","Atlantic/Reykjavik","Atlantic/South_Georgia","Atlantic/St_Helena","Atlantic/Stanley","Australia/Adelaide","Australia/Brisbane","Australia/Broken_Hill","Australia/Currie","Australia/Darwin","Australia/Eucla","Australia/Hobart","Australia/Lindeman","Australia/Lord_Howe","Australia/Melbourne","Australia/Perth","Australia/Sydney","Canada/Atlantic","Canada/Central","Canada/Eastern","Canada/Mountain","Canada/Newfoundland","Canada/Pacific","Europe/Amsterdam","Europe/Andorra","Europe/Astrakhan","Europe/Athens","Europe/Belgrade","Europe/Berlin","Europe/Bratislava","Europe/Brussels","Europe/Bucharest","Europe/Budapest","Europe/Busingen","Europe/Chisinau","Europe/Copenhagen","Europe/Dublin","Europe/Gibraltar","Europe/Guernsey","Europe/Helsinki","Europe/Isle_of_Man","Europe/Istanbul","Europe/Jersey","Europe/Kaliningrad","Europe/Kiev","Europe/Kirov","Europe/Lisbon","Europe/Ljubljana","Europe/London","Europe/Luxembourg","Europe/Madrid","Europe/Malta","Europe/Mariehamn","Europe/Minsk","Europe/Monaco","Europe/Moscow","Europe/Oslo","Europe/Paris","Europe/Podgorica","Europe/Prague","Europe/Riga","Europe/Rome","Europe/Samara","Europe/San_Marino","Europe/Sarajevo","Europe/Saratov","Europe/Simferopol","Europe/Skopje","Europe/Sofia","Europe/Stockholm","Europe/Tallinn","Europe/Tirane","Europe/Ulyanovsk","Europe/Uzhgorod","Europe/Vaduz","Europe/Vatican","Europe/Vienna","Europe/Vilnius","Europe/Volgograd","Europe/Warsaw","Europe/Zagreb","Europe/Zaporozhye","Europe/Zurich","GMT","Indian/Antananarivo","Indian/Chagos","Indian/Christmas","Indian/Cocos","Indian/Comoro","Indian/Kerguelen","Indian/Mahe","Indian/Maldives","Indian/Mauritius","Indian/Mayotte","Indian/Reunion","Pacific/Apia","Pacific/Auckland","Pacific/Bougainville","Pacific/Chatham","Pacific/Chuuk","Pacific/Easter","Pacific/Efate","Pacific/Enderbury","Pacific/Fakaofo","Pacific/Fiji","Pacific/Funafuti","Pacific/Galapagos","Pacific/Gambier","Pacific/Guadalcanal","Pacific/Guam","Pacific/Honolulu","Pacific/Kiritimati","Pacific/Kosrae","Pacific/Kwajalein","Pacific/Majuro","Pacific/Marquesas","Pacific/Midway","Pacific/Nauru","Pacific/Niue","Pacific/Norfolk","Pacific/Noumea","Pacific/Pago_Pago","Pacific/Palau","Pacific/Pitcairn","Pacific/Pohnpei","Pacific/Port_Moresby","Pacific/Rarotonga","Pacific/Saipan","Pacific/Tahiti","Pacific/Tarawa","Pacific/Tongatapu","Pacific/Wake","Pacific/Wallis","US/Alaska","US/Arizona","US/Central","US/Eastern","US/Hawaii","US/Mountain","US/Pacific","UTC"]
CheckSaveState = True #@param {type:"boolean"}
OptimizeColab = True #@param {type:"boolean"}
EnableSounds = True #@param {type:"boolean"}
UseOtherAccountAsMount = False #@param {type:"boolean"}
TextDelay = "Fast" #@param ["Normal", "Fast", "Instant"]

Delay = {"Normal" : 2,
         "Fast" : 1,
         "Instant" : .5
         }

bgcolor = {
    "black" : "\u001b[40m",
    "red" : "\u001b[41m",
    "green" : "\u001b[42m",
    "yellow" : "\u001b[43m",
    "blue" : "\u001b[44m",
    "magenta" : "\u001b[45m",
    "cyan" : "\u001b[46m",
    "white" : "\u001b[47m",
    "brightblack" : "\u001b[40;1m",
    "brightred" : "\u001b[41;1m",
    "brightgreen" : "\u001b[42;1m",
    "brightyellow" : "\u001b[43;1m",
    "brightblue" : "\u001b[44;1m",
    "brightmagenta" : "\u001b[45;1m",
    "brightcyan" : "\u001b[46;1m",
    "brightwhite" : "\u001b[47;1m",
    "end" : "\u001b[0m"

}

color = {
   "black" : "\u001B[30m",
   "purple" : "\033[95m",
   "cyan": "\033[96m",
   "darkcyan" : "\033[36m",
   "blue" : "\033[94m",
   "green" : "\033[92m",
   "yellow" : "\033[93m",
   "red" : "\033[91m",
   "bold" : "\033[1m",
   "underline" : "\033[4m",
   "reverse" : "\u001b[7m",
   "white" : "\u001b[37m",
   "brightblack" : "\u001b[30;1m",
   "brightred" : "\u001b[31;1m",
   "brightgreen" : "\u001b[32;1m",
   "brightyellow" : "\u001b[33;1m",
   "brightblue" : "\u001b[34;1m",
   "brightmagenta" : "\u001b[35;1m",
   "brightcyan" : "\u001b[36;1m",
   "brightwhite" : "\u001b[37;1m",
   "end" : "\033[0m"
   }

print(f"{color['bold']}{color['underline']}Initializing Environment{color['end']}:")
with output.use_tags('DriveAuthentication'):
    print(f"\n*Drive Authentication: {color['yellow']}Requesting.{color['end']}" , end="")
    print('\n')

    if not UseOtherAccountAsMount:
        from google.colab import drive
        drive.mount("/content/drive")
    else:    
        !add-apt-repository -y ppa:alessandro-strada/ppa 2>&1 > /dev/null
        !apt-get update -qq 2>&1 > /dev/null
        !apt-get -y install -qq google-drive-ocamlfuse fuse
        from google.colab import auth
        auth.authenticate_user()
        from oauth2client.client import GoogleCredentials
        creds = GoogleCredentials.get_application_default()
        import getpass
        !google-drive-ocamlfuse -headless -id={creds.client_id} -secret={creds.client_secret} < /dev/null 2>&1 | grep URL
        vcode = getpass.getpass()
        !echo {vcode} | google-drive-ocamlfuse -headless -id={creds.client_id} -secret={creds.client_secret}
        %cd /content
        !mkdir drive
        %cd drive
        !mkdir MyDrive
        %cd ..
        %cd ..
        !google-drive-ocamlfuse /content/drive/MyDrive

output.clear(output_tags='DriveAuthentication')
print(f"\n*Drive Authentication:{color['green']} Done{color['end']}")

with output.use_tags('SearchEngine'):
    print(f"\n*1337x Search Engine: {color['yellow']}Preparing.{color['end']}" , end="")
    print('\n')

    !pip install 1337x

output.clear(output_tags='SearchEngine')
from py1337x import py1337x
print(f"*1337x Search Engine:{color['green']} Done{color['end']}")


with output.use_tags('torPackage'):
    print(f"\n*Torrent Client Package: {color['yellow']}Preparing{color['end']}", end="")
    print('\n')

    !pip install python-libtorrent-bin #&> /dev/null
    !pip install magneturi #&> /dev/null

output.clear(output_tags='torPackage')
print(f"*Torrent Client Package: {color['green']}Done{color['end']}")
import os
import sys
import time
import shutil
import datetime
import ipywidgets as widgets
import threading
import textwrap
import pickle
import pytz


import libtorrent as lt
import magneturi

from IPython.display import display, clear_output
from datetime import datetime   
from google.colab import files
from google.colab import output
from threading import Thread

#######################Functions##################
class Sfx:
    @staticmethod
    def play(sfxType = ""):
        if EnableSounds:
            if sfxType == "intro":
                output.eval_js('new Audio("https://cdn.pixabay.com/download/audio/2022/02/22/audio_454e6d15e5.mp3?filename=warm-tech-logo-21474.mp3").play(),timeout_sec = 7')
            elif sfxType == "launch":
                output.eval_js('new Audio("https://cdn.pixabay.com/download/audio/2022/01/08/audio_4820a20bac.mp3?filename=simple-clean-logo-13775.mp3").play()',timeout_sec = 7)
            
            else:
                output.eval_js('new Audio("https://cdn.pixabay.com/download/audio/2021/08/04/audio_a233a6ff22.mp3?filename=level-win-6416.mp3",timeout_sec = 5).play()') #https://free-loops.com/data/mp3/42/f5/3b60ca3cdd98302d92a1aa941553.mp3

##########################UI######################
class UI:

    @staticmethod
    def bannerTorrentDownloader():
        print(f"")
        print(f" â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ•—      â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—    â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ•—       â–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ•—â–ˆâ–ˆâ•—      â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— ")
        print(f"â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘     â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—   â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•— â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•—  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘     â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â•â•â•â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—")
        print(f"â–ˆâ–ˆâ•‘  â•šâ•â•â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘     â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•¦â•   â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘ â•šâ–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•”â–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘     â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•")
        print(f"â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘     â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—   â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ–ˆâ–ˆâ•‘ â–ˆâ–ˆâ•‘â•šâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘     â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â•  â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—")
        print(f"â•šâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â•šâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•¦â•   â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â•šâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•  â•šâ–ˆâ–ˆâ•”â• â•šâ–ˆâ–ˆâ•”â• â–ˆâ–ˆâ•‘ â•šâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â•šâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘")
        print(f" â•šâ•â•â•â•â•  â•šâ•â•â•â•â• â•šâ•â•â•â•â•â•â•â•šâ•â•  â•šâ•â•â•šâ•â•â•â•â•â•    â•šâ•â•â•â•â•â•  â•šâ•â•â•â•â•    â•šâ•â•   â•šâ•â•  â•šâ•â•  â•šâ•â•â•â•šâ•â•â•â•â•â•â• â•šâ•â•â•â•â• â•šâ•â•  â•šâ•â•â•šâ•â•â•â•â•â• â•šâ•â•â•â•â•â•â•â•šâ•â•  â•šâ•â•")
        print(f"-Ì´Í„Í‹Í’ÌÍšÌ¢=ÌµÌ¾Í˜ÌºÍÌ¡-ÌµÍ›Ì½Ì¾Ì¼Ì¡Ì=ÌµÍŠÍŒÍÌªÌ™Íš-ÌµÍÍ›Í‹Í‰Ì¡Ì¢=Ì¸Ì¾ÍÌ¾Ì«Ì¢-Ì¸Í€Ì½ÌŸÌ˜Ì™=Ì¸ÍÍŒÌ“ÍœÍ‡ÌŸ-ÌµÌ•Ì’Í‘Ì¦ÌÌ =ÌµÍ†ÍÌ”Í”Ì¼Ì»-Ì´ÌšÌ•ÌšÍ‡Í“=Ì´Í‹Í€Ì½ÍÌÌŸ-Ì´Í‘ÌšÍ†Í™Ì«Í=Ì´Í„Ì’ÍŠÍšÌ¦Íœ-Ì¸ÌÍ€Í†Í‡ÍšÍš=Ì¸Í‹ÍÌ•Í–Í•-Ì¸Í„Í›Ì•ÌºÌª=Ì´Ì¾Ì”Ì™Ì¦Ì-Ì´Í„Í›Ì’Ì Ì¢Ì¼=Ì¸Í„Í†Ì¼ÌŸÌŸ-ÌµÍƒÍÌ½ÍœÌ»Ì˜=ÌµÍ„ÍÌ’ÍÌ¦Í“-Ì¸ÍÌ½Í•Ì»Í™=ÌµÍ‹Ì¾Í‘ÍÍÌª-Ì´Ì’Ì’Ì’Ì Ì™Íš=Ì¸ÍÍ‘Ì½ÌªÌ¦Ì»-Ì´Í„Ì¿Í‘Ì¡Í‡Ì¢=Ì´ÍÌ¾Í•ÌÌ -Ì´Ì½Ì“ÍŒÌ»Ì¢Ì¡=Ì¸Ì”Í‘ÌÌ¦-Ì´Ì¾Ì•Ì«Ì˜=Ì¸Í Ì•Ì¾Í“Í–Í”-ÌµÌ’Ì”Ì¡Ì™Ì»=Ì¸Í‹Í€Ì’Ì¦Í™Ì-Ì´ÍŒÌÍ›Í•Í”Í=Ì´ÍŠÌ¿Í–Ì¡Í™-Ì´ÍÍ„Í˜ÍÌŸ=Ì¸Ì¾ÍÍ‹Ì¢Ì»Í•-Ì´Í„Ì’ÍÌ¡Ì¦Íœ=Ì¸Í€Ì½ÍÍ™ÌªÌ-Ì¸Ì•Í„ÍÌ˜ÌÌ˜=Ì´Ì“ÍÍ’ÍšÌºÍ‡-Ì¸Í€Ì”ÍƒÍ“Í‡Ì¡=ÌµÌšÍ ÍŠÌ˜ÌºÌ¦-Ì´Í€ÍŒÍ•Ì˜Ì¢=Ì¸Í†ÌÌ™Í“Ì¼-Ì´Í„Í‹Í’ÌÍšÌ¢=ÌµÌ¾Í˜ÌºÍÌ¡-ÌµÍ›Ì½Ì¾Ì¼Ì¡Ì=ÌµÍŠÍŒÍÌªÌ™Íš-ÌµÍÍ›Í‹Í‰Ì¡Ì¢=Ì¸Ì¾ÍÌ¾Ì«Ì¢-Ì¸Í€Ì½ÌŸÌ˜Ì™=Ì¸ÍÍŒÌ“ÍœÍ‡ÌŸ-ÌµÌ•Ì’Í‘Ì¦ÌÌ =ÌµÍ†ÍÌ”Í”Ì¼Ì»-Ì´ÌšÌ•ÌšÍ‡Í“=Ì´Í‹Í€Ì½ÍÌÌŸ-Ì´Í‘ÌšÍ†Í™Ì«Í=Ì´Í„Ì’ÍŠÍšÌ¦Íœ-Ì¸ÌÍ€Í†Í‡ÍšÍš=Ì¸Í‹ÍÌ•Í–Í•-Ì¸Í„Í›Ì•ÌºÌª=Ì´Ì¾Ì”Ì™Ì¦Ì-Ì´Í„Í›Ì’Ì Ì¢Ì¼=Ì¸Í„Í†Ì¼ÌŸÌŸ-ÌµÍƒÍÌ½ÍœÌ»Ì˜=ÌµÍ„ÍÌ’ÍÌ¦Í“-Ì¸ÍÌ½Í•Ì»Í™=ÌµÍ‹Ì¾Í‘ÍÍÌª-Ì´Ì’Ì’Ì’Ì Ì™Íš=Ì¸ÍÍ‘Ì½ÌªÌ¦Ì»-Ì´Í„Ì¿Í‘Ì¡Í‡Ì¢â–ˆâ–€â–ˆ â–ˆâ–€ â–ˆâ–„â–ˆ â–ˆâ–€â–€ â–ˆ â–ˆâ€ƒâ–ˆâ–€â–ˆâ€ƒâ–ˆâ–€â–ˆâ€ƒâ–„â–€â–ˆâ€ƒâ–€â–ˆâ–€â€ƒâ–€â–ˆÍŠÌ˜ÌºÌ¦-Ì´Í€ÍŒÍ•Ì˜Ì¢=Ì¸Í†ÌÌ™Í“Ì¼")
        print(f"                                                                                          â–ˆâ–€â–€ â–„â–ˆ  â–ˆ  â–ˆâ–„â–„ â–ˆâ–€â–ˆâ€ƒâ–ˆâ–„â–ˆâ€ƒâ–ˆâ–€â–€â€ƒâ–ˆâ–€â–ˆâ€ƒ â–ˆ â€ƒâ–ˆâ–„ÍŠÌ˜ÌºÌ¦-Ì´")
    @staticmethod
    def bannerTorrentSearcher():
        print(f"â–€â–€â–ˆâ–€â–€ â–ˆâ–€â–€â–ˆ â–ˆâ–€â–€â–ˆ â–ˆâ–€â–€â–ˆ â–ˆâ–€â–€ â–ˆâ–€â–€â–„ â–€â–€â–ˆâ–€â–€ ã€€ â–’â–ˆâ–€â–€â–ˆ â–ˆâ–€â–€â–ˆ â–ˆâ–€â–€â–ˆ â–ˆâ–‘â–‘â–‘â–ˆ â–ˆâ–€â–€ â–ˆâ–€â–€ â–ˆâ–€â–€â–ˆ ")
        print(f"â–‘â–’â–ˆâ–‘â–‘ â–ˆâ–‘â–‘â–ˆ â–ˆâ–„â–„â–€ â–ˆâ–„â–„â–€ â–ˆâ–€â–€ â–ˆâ–‘â–‘â–ˆ â–‘â–‘â–ˆâ–‘â–‘ ã€€ â–’â–ˆâ–€â–€â–„ â–ˆâ–„â–„â–€ â–ˆâ–‘â–‘â–ˆ â–ˆâ–„â–ˆâ–„â–ˆ â–€â–€â–ˆ â–ˆâ–€â–€ â–ˆâ–„â–„â–€ ")
        print(f"â–‘â–’â–ˆâ–‘â–‘ â–€â–€â–€â–€ â–€â–‘â–€â–€ â–€â–‘â–€â–€ â–€â–€â–€ â–€â–‘â–‘â–€ â–‘â–‘â–€â–‘â–‘ ã€€ â–’â–ˆâ–„â–„â–ˆ â–€â–‘â–€â–€ â–€â–€â–€â–€ â–‘â–€â–‘â–€â–‘ â–€â–€â–€ â–€â–€â–€ â–€â–‘â–€â–€ ")
        print(f"                                                          ğŸ…¿ğŸ†‚ğŸ†ˆğŸ…²ğŸ…·ğŸ…¾ğŸ…¿ğŸ…°ğŸ†ƒğŸ†‚")
    
    @staticmethod
    def bannerFileManager():
        print(f"â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘")
        print(f"â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•—â–ˆâ–ˆâ•—â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–‘â–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–‘â–ˆâ–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–‘")
        print(f"â–‘â–ˆâ–ˆâ•”â•â•â•â•â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ•”â•â•â•â•â•â–‘â–‘â–‘â–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â•â•â•â–‘â–ˆâ–ˆâ•”â•â•â•â•â•â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–‘â–‘")
        print(f"â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ•”â–ˆâ–ˆâ–ˆâ–ˆâ•”â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â–ˆâ–ˆâ•—â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–‘â–‘")
        print(f"â–‘â–ˆâ–ˆâ•”â•â•â•â–‘â–‘â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ•”â•â•â•â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ•‘â•šâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â•šâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â•šâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•”â•â•â•â–‘â–‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–‘â–‘")
        print(f"â–‘â–ˆâ–ˆâ•‘â–‘â–‘â–‘â–‘â–‘â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–‘â–‘â–‘â–ˆâ–ˆâ•‘â–‘â•šâ•â•â–‘â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â•šâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â–‘â–‘â–ˆâ–ˆâ•‘â•šâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘â–‘â–‘â–ˆâ–ˆâ•‘â–‘â–‘")
        print(f"â–‘â•šâ•â•â–‘â–‘â–‘â–‘â–‘â•šâ•â•â•šâ•â•â•â•â•â•â•â•šâ•â•â•â•â•â•â•â–‘â–‘â–‘â•šâ•â•â–‘â–‘â–‘â–‘â–‘â•šâ•â•â•šâ•â•â–‘â–‘â•šâ•â•â•šâ•â•â–‘â–‘â•šâ•â•â•â•šâ•â•â–‘â–‘â•šâ•â•â–‘â•šâ•â•â•â•â•â•â–‘â•šâ•â•â•â•â•â•â•â•šâ•â•â–‘â–‘â•šâ•â•â–‘â–‘")
        print(f"â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘â–‘ğŸ…¿ğŸ†‚ğŸ†ˆğŸ…²ğŸ…·ğŸ…¾ğŸ…¿ğŸ…°ğŸ†ƒğŸ†‚â–‘")

    @staticmethod
    def display_msg(messageType,message):
        messageType = messageType.lower()
        msgType = ""
        colorType = ""
        bgColorType = ""
        if messageType == "error":
            msgType = "Error   "#5
            bgColorType = "red"
            colorType = "yellow"
        elif messageType == "success":
            msgType = "Success " #7
            bgColorType = "green"
            colorType = "purple"
        
        elif messageType == "warning":
            msgType = "Warning " #7
            bgColorType = "brightyellow"
            colorType = "purple"

        elif messageType == "check":
            msgType = "Checking"#8
            bgColorType = "yellow"
            colorType = "purple"
            
        elif messageType == "note":
            msgType  = "Note    "#4
            bgColorType = "cyan"
            colorType = "purple"
        else:
            msgType = "Missing MsgType"
            bgColorType = "red"
            colorType = color['purple']
        print(f"{bgcolor[bgColorType]}{color[colorType]} {msgType} {bgcolor['end']}{color['end']}: {message}")
        time.sleep(Delay[TextDelay])

    @staticmethod
    def set_userInput(messageConsole,message):
        print(f"{color['yellow']}{message}{color['darkcyan']}")
        userInput = input(f" {messageConsole}")
        print(f"{color['end']}")
        return userInput


    @staticmethod
    def display_items(Items): #Makeshift
        if Items:
            UI.display_chainBar(False,"darkcyan")
            for index,data in enumerate(Items[:]):
                UI.colorformatStart(index)
                print(f" [{'0' if index<10 else ''}{index+1}] {color['red']+'Invalid Item' if not data else data}")
                UI.colorformatEnd()
            UI.display_chainBar(True,"darkcyan")
        else:
            print(f"{color['end']}{color['darkcyan']}Currently Empty....{color['end']}")

    @staticmethod
    def set_verifyInput(messageConsole,message,choices):
        isRunning = True
        with output.use_tags('verifyInput'):
            while isRunning:
                userInput = UI.set_userInput(messageConsole,message).lower()
                if userInput in choices:
                    isRunning = False
                else:
                    UI.display_msg("error","Invalid choice!")

        output.clear(output_tags='verifyInput')
        return userInput


    @staticmethod
    def set_verifyInputInt(messageConsole,message,min,max):
        isRunning = True
        userInput = 0
        with output.use_tags('verifyInt'):
            while isRunning:
                userInput = UI.set_userInput(messageConsole,message)
                if not userInput.isnumeric():
                    UI.display_msg("error",f"The input must be an integer!")
                    continue
                
                userInput = int(userInput)

                if userInput >= min and userInput <= max :
                    isRunning = False
                else:
                    UI.display_msg("error",f"It must be between {min} - {max}") 
        output.clear(output_tags='verifyInt')
        return userInput

    @staticmethod
    def set_verifyInputBool(message):
        isRunning = True
        choice = True
        with output.use_tags('verifyBool'):
            userInput = UI.set_userInput(f"Yes/No: ",f"{color['yellow']}{message}{color['end']}")
            while isRunning:
                userInput = userInput.lower()
                if userInput in ['y','n','yes','no']:
                    isRunning = False
                    if userInput in ['n','no']:
                        choice = False
                else:
                    userInput = UI.set_userInput(f"Yes/No: ",f"{color['red']}Error{color['end']}:{color['yellow']}Invalid Input, Please choose only Y/N{color['end']}")
            
        output.clear(output_tags='verifyBool')
        return choice
  
    @staticmethod
    def set_verifyInputAdv(messageConsole,message,choices,ifHasIndex = False,ifsetOfIntegers = False,min=0,max=0):
        filter = choices
        if ifsetOfIntegers:
            isRunning = True
            while isRunning:
                userInput = UI.set_userInput(messageConsole,message)
                if not userInput.isnumeric():
                    UI.display_msg("error",f"The input must be an integer!")
                    continue
                if userInput >= min and userInput <= max :
                    isRunning = False
                
                UI.display_msg("error",f"The input must be between {min} - {max}")
        else:
            userInput = UI.set_userInput(messageConsole,message)
            if ifHasIndex and userInput.isnumeric():
                filter = filter + GeneralFunc.generate_enumArray(len(choices),False)
                userInput = choices[int(userInput)-1]
            else:
                userInput = userInput.lower()
            while userInput not in filter:
                userInput = UI.set_userInput(messageConsole,message)
                
                if ifHasIndex and userInput.isnumeric():
                    userInput = choices[int(userInput)-1]
                else:
                    userInput = userInput.lower()

        return userInput

    @staticmethod
    def colorformatStart(index):
        print(f"{color['end'] if index % 2 == 0 else color['darkcyan']}",end="")
    
    @staticmethod
    def colorformatEnd():
        print(color['end'],end="")

    @staticmethod
    def display_chainBar(ifEnd = False,barColor ="bold"):
        if ifEnd:
            print(f"{color['bold']}{color[barColor]}-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³=Ì³-Ì³{color['end']}")
        else:
            print(f"{color['bold']}{color[barColor]}-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…=Ì…Ì…-Ì…Ì…{color['end']}")

    @staticmethod
    def displayHelp(title,helpList): 
        print(f"{color['bold']}{color['yellow']}{title}:{color['end']}")
        UI.display_items(GeneralFunc.addListSymbol("info","blue",helpList))
        isRunning = True
        while isRunning:
            isRunning = not UI.set_verifyInputBool("Are you done Reading?")



####################Personalize##################
class Personalization:
    def __init__(self):
        self.infoLocation = TorrentHistory + "info.dat"
        self._username = ""
        self._totalDownloads = 0
        self._lastLogin = ""
        self._previousTotalDownloads = 0
        
    def setUsername(self,newName):
        self._username = newName

    def updateDownloadCounter(self):
        self._totalDownloads = self.getTotalDownloads()

    def getTotalDownloads(self):
        return len(GeneralFunc.getItemsFromDisk(TorrentHistory+"Completed",".torrent"))

    def getUsername(self):
        return self._username

    def saveRecentInfo(self):
        data = [self.getUsername(),self._totalDownloads,GeneralFunc.timeCheck(False)]
        if DEBUG:
            print(f"SaveRecent data = {data}")
            input
        GeneralFunc.saveFromDisk(self.infoLocation,data)

    def getRecentInfo(self):
        data = GeneralFunc.readFromDisk(self.infoLocation)
        if DEBUG:
            print(f"RecentInfo data = {data}")
        return data

    def getPartOfTheDay(self):
        currentHour = int(datetime.now(pytz.timezone(Timezone)).strftime("%H"))
        if currentHour == 0 :
            return "Midnight"
        elif currentHour >= 1  and currentHour < 5:
            return "Late Night"
        elif currentHour >= 5 and currentHour < 8:
            return "Early Morning"
        elif currentHour >= 8 and currentHour < 12:
            return "Morning"
        elif currentHour == 12:
            return "Noon"
        elif currentHour >= 13 and currentHour < 18:
            return "Afternoon"
        elif currentHour >= 18 and currentHour < 24:
            return "Evening"

    def greeter(self):
        partOfDay = self.getPartOfTheDay()
        if  partOfDay == "Evening":
            return "Evening"
        elif partOfDay in ["Midnight", "Early Morning", "Morning", "Late Night"]:
            return "Morning"
        elif partOfDay in ["Noon","Afternoon"]:
            return "Afternoon"
        else:
            return "Error"

    def inputNewName(self):
        isRunning = True
        data = ""
        while isRunning:
            with output.use_tags('inputNewName'):
                data = UI.set_userInput("New Username: ",f"{'Welcome New User, Please Input your name.' if not self.getUsername() else 'Hi '+ self.getUsername() +' what would your new name be like?'}")
                if UI.set_verifyInputBool(f"{'Are you sure your name is '+ data if not self.getUsername() else 'Are you sure you want to change your name as '+ self.getUsername() +'?'}"):
                    UI.display_msg("success",f"Successfully {'set'  if not self.getUsername() else 'changed'} your Username, Welcome {self.getUsername()}!")
                    self.setUsername(data.capitalize())
                    isRunning = False
                    continue
                if self.getUsername() and UI.set_verifyInputBool(f"Continue changing your name? {self.getUsername()}"):
                    isRunning = False
            output.clear(output_tags='inputNewName')  


    def updateInfo(self):
        with output.use_tags('nameInit'):
            data = self.getRecentInfo()
        output.clear(output_tags='nameInit')
        if not data:
            with output.use_tags('nameInit'):
                UI.display_msg("warning","No recent Userdata found. Now Creating..")
                self.inputNewName()
            output.clear(output_tags='nameInit')
            UI.display_msg("success","User Initialization Done..")
            self._lastLogin = GeneralFunc.timeCheck(False)
        else:
            self.setUsername(data[0]) 
            self._previousTotalDownloads = data[1]
            try:
                self._lastLogin = data[2]
            except IndexError:
                self._lastLogin = GeneralFunc.timeCheck(False)
            if DEBUG:
                print(data)
                print(self._lastLogin)
                input()

        self.updateDownloadCounter()
        self.saveRecentInfo()

    def greetUser(self):
        self.updateInfo()
        daysPassed = GeneralFunc.getDaysTimePast(GeneralFunc.timeCheck(False),self._lastLogin)
        userStats = [f"Total Torrent Downloads: {self._totalDownloads}",f"Last Session Newly Added Items: + {self._totalDownloads - self._previousTotalDownloads} Items",f"Total Unfinished Downloads: {len(File_Manager.get_Unfinished())} Items",f"Last Session Date: {self._lastLogin} {'this day.' if not daysPassed else str(daysPassed) + ' day ago.' if daysPassed == 1 else str(daysPassed) + ' days ago.'} "]
        print(f"\n{color['yellow']}Welcome and Good {self.greeter()}! {color['bold']}{color['green']}{self.getUsername()}{color['end']}.")
        print(f"{color['yellow']}Have a great {self.getPartOfTheDay()}!{color['end']}")
        print(f"\n{color['bold']}{color['yellow']}Stats:{color['end']}")
        UI.display_items(userStats)


        
#$$$$$$$$$$$$$GeneralFunc##############

class GeneralFunc:
    @staticmethod
    def addListSymbol(insertedSymbol,symbolColor,listItems):
        newList = []
        for data in listItems:
            newList.append(f"{color[symbolColor]}{symbol[insertedSymbol]}{color['end']}{data}")
        return newList

    @staticmethod
    def generate_enumArray(max,ifStartWithZero = True):
        container = []
        shift = 1
        if not ifStartWithZero:
            shift = 0

        while max:
            container.append(max - shift)
            max -= 1
        return container

    @staticmethod
    def addExitToDict(items):
        array = []
        for data in items:
            array.append(data)
        array.append("exit")
        return array

    @staticmethod
    def timeCheck(ifHourly= True):
        timeChecked = datetime.now(pytz.timezone(Timezone))
        if ifHourly: 
            return timeChecked.strftime("%I:%M %p")
        return timeChecked.strftime("%m/%d/%Y")

    @staticmethod
    def removeFromList(Items,data):
        Items.remove(data)

    @staticmethod
    def addFromList(Items,data):
        Items.append(data)
        
    @staticmethod
    def removeFromDisk(location):
        os.remove(location)
    
    @staticmethod
    def checkFromDisk(location):
        ifExist = False
        if os.path.exists(location):
            ifExist = True
        return ifExist

    @staticmethod
    def saveFromDisk(location,data):
         with open(location, 'wb') as fp:
            pickle.dump(data, fp)
    
    @staticmethod
    def readFromDisk(location):
        try:
            with open (location, 'rb') as fp:
                data = pickle.load(fp)
        except FileNotFoundError:
            UI.display_msg("error","File not found.")
            return False

        except EOFError:
            UI.display_msg("error","File is currupted,")
            return False

        return data
    
    @staticmethod
    def getItemsFromDisk(location,ifEndsWith = "",):
        paths = os.listdir(location)
        if ifEndsWith:
            paths = [path for path in paths if path.endswith(ifEndsWith)]
        return paths


    @staticmethod
    def get_rate(val):
        prefix = ['B', 'kB', 'MB', 'GB', 'TB']
        for i in range(len(prefix)):
            if abs(val) < 1000:
                if i == 0:
                    return '%5.3g %s' % (val, prefix[i])
                else:
                    return '%4.3g %s' % (val, prefix[i])
            val /= 1000

        return '%6.3g PB' % val

    @staticmethod
    def get_eta(fileSize,speed):
        try:
            eta = fileSize / speed

        except ZeroDivisionError:
            return 0
            
        return eta 

    @staticmethod
    def get_secToTime(seconds):
        if seconds == 0:
            return f"."
        seconds = int(seconds)
        hh = str(seconds//3600)
        mm = str((seconds%3600)//60)
        ss = str((seconds%3600)%60)
        message = ""
        
        if int(hh) > 24:
            message = f"{int(int(hh)/ 24)} days."
        
        elif int(hh) :
            message = f"{hh}:{'0' if int(mm) < 10 else ''}{mm}:{ss}"
        else:
            message = f"{'' if int(hh) == 0 else hh+ ' hours '}{'' if int(mm) == 0 else mm + 'm '}{'' if int(ss) == 0 else ss + 's.'}"
        
        return message


    def getDaysTimePast(date1,date2):
        date_format = "%m/%d/%Y"
        if DEBUG:
            print(f"date1 = {date1}\nDate2 = {date2}")
            input()
        b = datetime.strptime(date1, date_format)
        a = datetime.strptime(date2, date_format)
        delta = b - a
        return delta.days

    @staticmethod
    def transferTorrent(torrentName,transferType = "complete"):
        if transferType == "complete":
            try:
              shutil.move(TorrentHistory + torrentName + '.torrent', completed_path + torrentName + '.torrent')
            except:
              pass
        elif transferType == "delete":
            try:
              shutil.move(TorrentHistory + torrentName, deleted_path + torrentName)
            except:
              UI.display_msg("error",f"{torrentName} {color['yellow']} not found{color['end']}")
        else:
            UI.display_msg("error","Unknown Location")
          

######################Loading Progress####################
class LoadingSpinner:
    busy = False
    delay = 0.3

    @staticmethod
    def spinning_cursor():
        while 1: 
            for cursor in '|/-\\': yield cursor

    def __init__(self, delay=None):
        self.spinner_generator = self.spinning_cursor()
        if delay and float(delay): self.delay = delay

    def spinner_task(self):
        while self.busy:
            sys.stdout.write(next(self.spinner_generator))
            sys.stdout.flush()
            time.sleep(self.delay)
            sys.stdout.write('\b')
            sys.stdout.flush()

    def __enter__(self):
        self.busy = True
        threading.Thread(target=self.spinner_task).start()

    def __exit__(self, exception, value, tb):
        self.busy = False
        time.sleep(self.delay)
        if exception is not None:
            return False


##########################Main Functions########################################################
class MainFunc:
   
    @staticmethod
    def set_addTorrent(magnetLink):
        downloads.append(
            lt.add_magnet_uri(ses, magnetLink, params)
        )
    
    @staticmethod
    def set_verifyTorrentImport(magnetLink,ifSaveToHistory = True,addToQueue = True):
        if MainFunc.process_magnet(magnetLink,ifSaveToHistory):
            if addToQueue:
                MainFunc.set_addTorrent(magnetLink)
            return True
        return False
    
    @staticmethod
    def set_inputMagnet():
        isRunning = True
        magnetLink = ""
        while isRunning:
            clear_output()
            UI_DownloadMenu.display_downloadQueue()
            magnetLink = UI.set_userInput("Input: ", f"{color['yellow']}Enter {color['bold']}Magnet Link{color['exit']}{color['yellow']} Or Type {color['bold']}Exit{color['exit']}{color['yellow']}: ")
            if magnetLink.lower() == "exit":
                isRunning = False
            MainFunc.set_verifyTorrentImport(magnetLink)
        return

    @staticmethod
    def ifDuplicateDownloadsFound(data):
        found = False
        for index, download in enumerate(downloads[:]):
            if download.name() == data:
                found = True
                UI.display_msg("error",f"Torrent Already Existed on download Index: {index +1}. File: {color['underline']}{download.name()[:40]}...{color['end']}")
                break
        for item in finished:
            if item == data:
                found = True
                UI.display_msg("error",f"Torrent Already Downloaded. File: {color['underline']}{data[:50]}...{color['end']}")
                break
        return found

    @staticmethod
    def set_importTorrent():
        try:
            while True:
                clear_output() 
                UI_DownloadMenu.display_downloadQueue()
                source = files.upload()
                filename = next(iter(source))
                if MainFunc.process_torrent(filename) == "NULL":
                    del source
                    break

                magnetLink = MainFunc.process_torrent(filename)
                MainFunc.set_verifyTorrentImport(magnetLink)
                del source
        except StopIteration:
            return
            

    @staticmethod
    def remove_torrent(download):
        ses.remove_torrent(download)
        downloads.remove(download)

    @staticmethod
    def remove_torrentDownload():
        clear_output()
        i = 0
        if not downloads:
            UI.display_msg("error",f"Download Queue is {color['red']}Empty{color['end']}, there's nothing to remove.")
        else:
            print(f"{color['yellow']}Download Queue:{color['end']}{color['end']}{color['darkcyan']} ")
            UI_DownloadMenu.display_downloadQueue()
            print(f"{color['end']}") 
            choice = UI.set_verifyInputInt("Index: ","Choose what index to remove:",1,len(downloads))
            choice -= 1
            notFound = True
            for index, download in enumerate(downloads[:]):
                if choice == index:   
                    UI.display_msg("success",f"{color['underline']}{download.name()}{color['cyan']}Removed{color['end']}")
                    GeneralFunc.addFromList(Logs,f"   {color['red']}{symbol['cross']}{color['end']}{color['red']} Removed{color['end']}: {download.name()} is {color['cyan']}removed{color['end']} on {color['cyan']}{GeneralFunc.timeCheck()}{color['end']}")
                    notFound = False
                    MainFunc.remove_torrent(download)
                    break
            if notFound:  
                UI.display_msg("error",f"Torrent index[{choice+1}] is {color['red']}not found{color['end']}")
        
                        

    @staticmethod
    def process_torrent(file):
        try:
            magnet = magneturi.from_torrent_file(file)
        except SyntaxError:
            UI.display_msg("error",f"{color['end']}{file} is not a valid bittorent file. ")
            magnet = "NULL"
            return magnet
        
        except AssertionError:
            if os.path.exists(file):
                UI.display_msg("error",f"{file} is already Completed therefore not added to queue. ")
            else:
                UI.display_msg("error",f"{file} not Found.")
            magnet = "NULL"
            return magnet

        except : #Most Commonly DecodingException
            UI.display_msg("error",f"Torrent is currupted, please download it again. Possible causes {color['yellow']}google drive quota exceeded.{color['end']} Filename: {file}")
            magnet = "NULL"
        
        return magnet


    @staticmethod
    def process_magnet(magnetLink,ifSaveToHistory = True):
        try:
            params = lt.parse_magnet_uri(magnetLink)
        except:
            UI.display_msg("error",f"Only valid Magnet link is accepted!")
            return False

        session = lt.session()
        handle = session.add_torrent(params)
        timer = 0
        ifInvalid = False
        print(f"{color['red']}{bgcolor['yellow']} Checking {bgcolor['end']}: {color['green']}Getting torrent Metadata {color['end']}", end='')
        with LoadingSpinner():
            while not handle.has_metadata(): 
                timer = timer + 1
                time.sleep(0.1)
                if timer == 650:
                    print(f"\b, It\'s slow due to {color['red']}Low Seeders {color['end']}")
                if timer >= 700: #1 min estimation
                    ifInvalid = True
                    break
                
                    
        if ifInvalid:
            UI.display_msg("error",f"{color['red']}Timeout Error{color['end']}, can\'t download torrent metadata. Try again next time.")
            return False 

        torrent_info = handle.get_torrent_info()
        if MainFunc.ifDuplicateDownloadsFound(torrent_info.name()):
            return False

        if ifSaveToHistory:
            torrent_file = lt.create_torrent(torrent_info)
            torrent_path = os.path.join(TorrentHistory, torrent_info.name() + ".torrent")
          
            with open(torrent_path, "wb") as f:
                f.write(lt.bencode(torrent_file.generate()))
            UI.display_msg("success",f"{'This Torrent' if not torrent_info.name() else torrent_info.name()} is {color['green']}Added")
        return True
    
    @staticmethod
    def set_addFileToQueue(fileName,ifCheckMetadata):
        file = TorrentHistory + fileName
        magnetLink = MainFunc.process_torrent(file)
        if magnetLink == "NULL":
            return False
        if ifCheckMetadata:
            UI.display_msg("check",f"{fileName} ")
            if MainFunc.set_verifyTorrentImport(magnetLink,False):
                UI.display_msg("success",f"Added to download queue: {color['underline']}{fileName[:40]}...{color['end']}")
                return True
        else:
            fileName = fileName.split(".torrent")
            if not MainFunc.ifDuplicateDownloadsFound(fileName[0]):
                MainFunc.set_addTorrent(magnetLink)
                UI.display_msg("success",f"Added to queue: {fileName[0]}")
                return True
            else:
                return False

    @staticmethod
    def recentSaveUpdater():
        downloadQueue = []
        saveLocation = TorrentHistory + 'recents.dat'
        
        for index, download in enumerate(downloads[:]):
            GeneralFunc.addFromList(downloadQueue,download.name())

        if not downloadQueue:
            if GeneralFunc.checkFromDisk(saveLocation):
                GeneralFunc.removeFromDisk(saveLocation)
        
        else:
            GeneralFunc.saveFromDisk(saveLocation,downloadQueue)



    @staticmethod
    def recentSaveLoader():
        saveLocation = TorrentHistory + 'recents.dat'
        invalid = 0

        if not downloads:
            downloadQueue = []
            UI.display_msg("check",f"Searching for Previous Unfinished queue:")
            if GeneralFunc.checkFromDisk(saveLocation):
                downloadQueue = GeneralFunc.readFromDisk(saveLocation)
                
                if not downloadQueue:
                    return
                
                for data in downloadQueue:
                    if not data:
                            invalid += 1
                if invalid != 0:
                    if invalid == len(downloadQueue):
                        UI.display_msg("error",f"{color['bold']}All Recent Downloads are invalid.")
                    else:
                        UI.display_msg("warning",f"{color['bold']}There are {color['yellow']}{invalid}{color['end']} invalid download {'item'if invalid == 1 else 'items'} found,Possible reason is the program didn't download the name metadata yet. \nPlease use the {color['yellow']}Torrent Manager{color['end']}{color['bold']} in the main menu to redownload it.")
                  
            else:
                UI.display_msg("note",f"No recent savedata found.{color['end']}")
                return  
                      
            userInput = ""

            with output.use_tags('getRecentUnfinished'):
                if len(downloadQueue) > 0:
                    UI.display_msg("success",f"A previous download list is recovered:")
                    UI.display_items(downloadQueue)
                    userInput = UI.set_verifyInputBool("Do you wanna resume downloading these?")
                if userInput:
                    for data in downloadQueue:
                        if not data:
                            continue
                        MainFunc.set_addFileToQueue(f"{data}.torrent",False)
                else:
                    UI.display_msg("warning",f"Found unfinished download queue {color['cyan']}discarded{color['end']}.")
            output.clear(output_tags='getRecentUnfinished')
            print(f"\b {color['green']}Done")

#################@UI Download Menu#################

class UI_DownloadMenu:
    
    @staticmethod
    def process_choice(choice):
        if choice == 0:
            return
        elif choice == 1:
            SearchEngine = TorrentSearch()
            SearchEngine.main()
            del SearchEngine             
        elif choice == 2:
            MainFunc.set_inputMagnet()
        elif choice == 3:
            MainFunc.set_importTorrent()
        elif choice == 4:
            x = File_Manager()
            x.main()
            del x

        
        
    @staticmethod
    def display_downloadQueue():
        if downloads:
            print (f"\n{color['green']}Download Queue{color['end']}:")
            for index, download in enumerate(downloads[:]):
                print (f"[{'0' if index < 10 else ''}{index + 1}] {download.name()}, Seeders:{color['green'] if download.status().num_seeds > 100 else color['yellow'] if download.status().num_seeds > 30 else color['red']}{download.status().num_seeds}{color['end']}{color['darkcyan']}(Not Realtime){color['end']}")
        else:
            print(f"{color['yellow']}Download Queue is currently {color['darkcyan']}Empty{color['end']}!")

    @staticmethod
    def display_logs():
        
        if Logs or downloads:
            print(f"{color['yellow']}\nRecent Logs{color['end']}:")
            if Logs:
                for data in Logs:
                    print(f"{data}")
        elif not downloads and not Logs:
            print(f"{color['yellow']}\nRecent Logs{color['end']}:{color['bold']} Currently Empty...{color['end']}")
            

    @staticmethod
    def display_menu():
        UI.display_chainBar(True)
        print(f"Menu:")
        print(f" [{color['bold']}{color['underline']}1{color['end']}]{color['cyan']}{color['underline']}Torrent Browser{color['end']} | [{color['bold']}{color['underline']}2{color['end']}]{color['cyan']}{color['underline']}Import via Magnet{color['end']} | [{color['bold']}{color['underline']}3{color['end']}] {color['cyan']}{color['underline']}Import via .torrent{color['end']} | [{color['bold']}{color['underline']}4{color['end']}]{color['cyan']}{color['underline']}File Manager{color['end']} | [{color['bold']}{color['underline']}0{color['end']}]{color['cyan']}{color['underline']}Exit{color['end']}\n")
        
    
    @staticmethod
    def main():
        isRunning = True
        UI_DownloadMenu.display_menu()
        choice = UI.set_verifyInputInt("Choice: ","Enter Choice: ",0,4)
        UI_DownloadMenu.process_choice(choice)


    def header():
        print(f"{color['bold']}{color['yellow']}Download Queue:{color['end']}")
        if not downloads:
            if not finished:
                print(f"{color['bold']}No Torrent Found, Please add one in the menu.\n{color['end']}")

    def set_inputMenu():
        choice = ""
        with output.use_tags('menu'):
            choice = input("[Default]Menu [/Search \"Keyword\"] \t [R]Remove Torrent \t [Q]Quit\nEnter Magnet\\Choice: ")
        output.clear(output_tags='menu')
        print("\n")
        return choice

    def ifMagnet(data):
        isMagnet = False 
        if len(data) >= 11:
            if data[:11] == "magnet:?xt=":
                isMagnet = True
        return isMagnet

    def ifSearch(data):
        isSearch =False
        if len(data) >= 7:
            if str(data[:7]).lower() == "/search":
                isSearch = True
        return isSearch

    def processIfSearch(data):
        index = 8
        if len(data)<8:
            index = 7
        SearchEngine = TorrentSearch()
        SearchEngine.main(data[index:])
        del SearchEngine


    
    def processIfMagnet(data):
        UI.display_chainBar()
        UI.display_msg("check",f"{color['yellow']}Processing Magnet: {color['darkcyan']}{color['underline']}{data[:8]}...{data[64:]}{color['end']}")
        MainFunc.set_verifyTorrentImport(data)

            
########################File Manager#####################
class File_Manager:

    def __init__(self):
      self._currentQueue = self.get_Unfinished()
      self._selectedDownloadQueue = []
      self._running = True

    def kill(self):
        self._running = False

    def add_toQueue(self,items,data):
        GeneralFunc.addFromList(items,data)

    def remove_toQueue(self,items,data):
        GeneralFunc.removeFromList(items,data)

    def delete_torrentFile(self, data):
        UI.display_msg("success",f"{color['yellow']}Removed: {color['darkcyan']}{data}{color['end']}")
        GeneralFunc.transferTorrent(data,"delete")

    @staticmethod
    def get_Unfinished():
        return GeneralFunc.getItemsFromDisk(TorrentHistory,'.torrent')

    def displayQueue(self,items):
        UI.display_items(items)
    
    def selectAllList(self):
        self._selectedDownloadQueue += self._currentQueue
        self._currentQueue = []

    def restoreDefaultList(self):
        self._currentQueue = File_Manager.get_Unfinished()
        self._selectedDownloadQueue = []
      
    def updateQueue(self,data, updateType):
        if updateType == "remove":
            self.add_toQueue(self._currentQueue,data)
            self.remove_toQueue(self._selectedDownloadQueue,data)
        elif updateType == "add":
            self.add_toQueue(self._selectedDownloadQueue,data)
            self.remove_toQueue(self._currentQueue,data)
        else:
            UI.display_msg("error","Unknown updateType") #for redundancy

          
    def clear_unfinishedTorrent(self):
        if not self._currentQueue:
           UI.display_msg("warning","No recent downloads found.")
        else:
            choice = UI.set_verifyInputBool("Are you sure you want to clear recents? ")
            UI.display_msg("note",f"{color['yellow']}These files will be moved in Deleted folder{color['end']}")
            if choice:
                for data in self._currentQueue:
                    self.delete_torrentFile(data) 
            time.sleep(Delay[TextDelay]) 

    def menu_removeFromQueue(self):
        clear_output()
        if self._selectedDownloadQueue:
            print(f"{color['yellow']}Selected Items:{color['end']}{color['cyan']}")
            self.displayQueue(self._selectedDownloadQueue)
            print(color['end'])
            choice = UI.set_verifyInputInt("Choice: ","Input index to remove: ",1,len(self._selectedDownloadQueue))
            self.updateQueue(self._selectedDownloadQueue[choice - 1],"remove")
        else:
            UI.display_msg("note",f"There's nothing to remove...")

    def finalize(self):
        if len(self._selectedDownloadQueue):
            print(f"{color['yellow']}Do you want to verify {'these torrents' if len(self._selectedDownloadQueue) > 1 else 'this torrent'} for seeders?{color['darkcyan']}(May take longer){color['end']}")
            checkMetadata = UI.set_verifyInputBool("Check for Verification?")
            invalid = 0
            for data in self._selectedDownloadQueue:
                UI.display_msg("note",f"{color['yellow']}File:{color['end']} {color['underline']}{color['cyan']}{data}{color['end']}")
                if not MainFunc.set_addFileToQueue(data,checkMetadata):
                    invalid = invalid + 1
                else:
                    UI.display_msg("success",f"{color['green']} Added{color['end']} {color['underline']}{color['cyan']}{data}{color['end']}")
            UI.display_msg("success",f"{color['yellow']}Import complete. Errors found: {invalid} {color['end']}")
        else:
            UI.display_msg("note",f"No data is imported")
    def processInput(self,userInput):
        ifValid = False
        if userInput.isnumeric():
            userInput = int(userInput) - 1
            if userInput >= 0 and userInput < len(self._currentQueue):
                self.updateQueue(self._currentQueue[userInput],"add")
                ifValid = True
            else:             
                UI.display_msg("error",f"Input must not exceed 0-{len(self._currentQueue)}")
                
        return ifValid
             
    def main(self):
        clear_output()
        if len(self._currentQueue): 
            while self._running:
                clear_output(True)
                UI.bannerFileManager()
                print(f"{color['yellow']}Enter the Index to be added to the queue.{color['end']}")
                print(f"{color['yellow']}Found Unfinished Items:{color['end']}{color['darkcyan']}")
                self.displayQueue(self._currentQueue)
                print(f"{color['yellow']}Selected Items:{color['end']}{color['cyan']}")
                self.displayQueue(self._selectedDownloadQueue)
                print(f"{color['end']}\nChoice: [{color['underline']}A{color['end']}]Select All [{color['underline']}R{color['end']}]Remove from Queue [{color['underline']}D{color['end']}]Return Default List [{color['underline']}C{color['end']}]Clear all Unfinished [{color['underline']}X{color['end']}]Exit")
                print(f"{color['yellow']}Choice: Enter Choice:{color['end']}")
                userInput = input("Choose: ")
                    
                if not userInput:
                    UI.display_msg("error","Please Input something.")
                    continue
                if self.processInput(userInput):
                    time.sleep(Delay[TextDelay])
                    continue
                elif userInput.lower() == 'r':
                    self.menu_removeFromQueue()
                elif userInput.lower() == 'd':
                    self.restoreDefaultList()
                elif userInput.lower() == 'c':
                    self.clear_unfinishedTorrent()
                elif userInput.lower() == 'a':
                    self.selectAllList()
                elif userInput.lower() == 'x':
                    self.finalize()
                    time.sleep(Delay[TextDelay])
                    self.kill()

        else:
            print(f"There's currently no unfinished download at the moment. Now go get some..")
            time.sleep(Delay[TextDelay])


######################@title Start Program################
class UI_Download:

    
    def __init__(self):
        self._running = True
        self._animationFrames = 0
        self._nextShift = 0
    
    # Thread Killing
    def kill(self):
        self._running = False
    def defaultNextShift(self):
        self._nextShift = 0
    def updateNextShift(self):
        self._nextShift += -1

    def get_nextShift(self):
        return self._nextShift
    
    def updateAnimFrames(self):
        if self._animationFrames >= 2:
            self._animationFrames = 0
        else: 
            self._animationFrames += 1

    def get_animationFrames(self):
        return self._animationFrames

    # Print Status Bar
    def completeDownload(self,name):
        currentMessage = f"   {color['yellow']}{symbol['star']}{color['end']}{color['green']} Completed{color['end']}: {name} {color['cyan']}completed{color['end']} on {color['cyan']}{GeneralFunc.timeCheck()}{color['end']}"
        print(currentMessage)
        GeneralFunc.addFromList(Logs,currentMessage)
        GeneralFunc.addFromList(finished,str(name))
        GeneralFunc.transferTorrent(name,"complete")


    def show(self) :
        MainFunc.recentSaveUpdater()
        layout = widgets.Layout(width="auto")
        style = {"description_width": "initial"}
        download_bars = [
            widgets.FloatSlider(
                step=0.01, disabled=True, layout=layout, style=style
            )
            for _ in downloads
        ]
        display(*download_bars)
        while self._running and downloads:
            self.defaultNextShift()
            for index, download in enumerate(downloads[:]):
                bar = download_bars[index + self.get_nextShift()]
                status = download.status()
                if not download.is_seed():
                    bar.value = status.progress * 100
                    bar.description = " ".join(
                        [
                            f"\t",
                            f"{symbol[state_str[status.state]]}",
                            f"{download.name()[:28]}...",
                            f" | {seedQuality['high'] if status.num_seeds >= 100 else seedQuality['med'] if status.num_seeds >=30 else seedQuality['low'] }Seeder:{status.num_seeds}",
                            f" | {animDownload[self.get_animationFrames()]} {GeneralFunc.get_rate(status.download_rate)}/s",
                            f" | {animUpload[self.get_animationFrames()]} {GeneralFunc.get_rate(status.upload_rate)}/s",
                            f" | Size:{GeneralFunc.get_rate(status.total_wanted_done)}/{GeneralFunc.get_rate(status.total_wanted)}",
                            f"{'| '+ symbol['time'] if state_str[status.state] == 'Downloading' else ''} {GeneralFunc.get_secToTime(GeneralFunc.get_eta(status.total_wanted - status.total_wanted_done,status.download_rate)) if state_str[status.state] == 'Downloading' else '.'}",
   
                        ]
                    )
                
                else:
                    self.completeDownload(download.name())
                    self.updateNextShift()
                    MainFunc.remove_torrent(download)
                    MainFunc.recentSaveUpdater()
                    bar.close()
                    download_bars.remove(bar)
                    Sfx.play()
                    
                    
                self.updateAnimFrames()
            time.sleep(.5)
            
        
        if finished and not downloads:
            print(f"\n{symbol['checkmark']} {color['green']}All Pending Downloads are Completed, Consider adding some!{color['end']}")


#Global Vars
state_str = [
    "Queued",
    "Checking",
    "Downloading metadata",
    "Downloading",
    "Finished",
    "Seeding",
    "Allocating",
    "Checking fastresume",
]

symbol = {
    "Queued" : "â§—",
    "Checking" : "â™»",
    "Downloading metadata" :"âš",
    "Downloading" : "ğŸ¡‡",
    "Finished" : "âˆ",
    "Seeding" : "â†¹",
    "Allocating" : "Ö",
    "Checking fastresume" :"â—Œ",
    
    "checkmark" : "âœ…",
    "star" : "âœ¯",
    "seeder" : "â—•",
    "speed" : "â‡…",
    "file" : "â™",
    "warning": "âš ",
    "skull" :"ğŸ•±",
    "arrowUp" : "â¯­",
    "arrowDown" : "â¯¯",
    "time" : "â˜¯",
    "cross" : "â¨³",
    "info" : "ğŸ›ˆ"
} 

seedQuality = {
    "low" : "â™¦ï¸",
    "med" : "ğŸ”¶",
    "high" : "ğŸŸ¢"
}

animDownload = {
    1 : "ğŸ¡£",
    0 : "ğŸ¡³",
    2 : "ğŸ¢ƒ"

}

animUpload = {
    2 : "ğŸ¡¡",
    1 : "ğŸ¡±",
    0 : "ğŸ¢"
}


searchSortBy = {
    "time" : "time",
    "size" : "size",
    "seeders" : "seeders",
    "leechers" : "leechers",
}

searchSortOrder = {
    "descending" : "desc",
    "ascending" : "asc"
}

if not Downloads:
    Downloads = "/content/drive/MyDrive/Downloads"
    UI.display_msg("warning",f"Download location is empty, reseting to default.{Downloads}")

SaveDirectory = Downloads + '/MyTorrents' 
TorrentHistory = Downloads + '/TorrentHistory/'
completed_path = TorrentHistory + 'Completed/'
deleted_path = TorrentHistory + 'Deleted/'

params = { ##https://www.rasterbar.com/products/libtorrent/single-page-ref.html
    "save_path": SaveDirectory,
          }

torrentSettings = { ##https://libtorrent.org/reference-Settings.html#settings_pack
  'connection_speed' : 4000,
  'connections_limit' : 500,
  'active_downloads' : 8,
  'active_checking' : 8}
ses = lt.session(torrentSettings)
ses.listen_on(6881, 6891)

downloads = []
finished = []
Logs = []

for path in [TorrentHistory, SaveDirectory, completed_path]:
 if not os.path.exists(path):
	  os.makedirs(path) 

x = Personalization()
x.greetUser()
UI.display_msg("note","Now run the next cell to continue...")
del x
Sfx.play("intro")

```

```python
class Personalization:
    def __init__(self):
        self.infoLocation = TorrentHistory + "info.dat"
        self._username = ""
        self._totalDownloads = 0
        
    def setUsername(self,newName):
        self._username = newName

    def updateDownloadCounter(self):
        self._totalDownloads = self.getTotalDownloads()

    def getTotalDownloads(self):
        return len(GeneralFunc.getItemsFromDisk(TorrentHistory+"Completed",".torrent"))

    def getUsername(self):
        return self._username

    def saveRecentInfo(self):
        data = [self.getUsername(),self._totalDownloads]
        if DEBUG:
            print(f"SaveRecent data = {data}")
        GeneralFunc.saveFromDisk(self.infoLocation,data)

    def getRecentInfo(self):
        data = GeneralFunc.readFromDisk(self.infoLocation)
        if DEBUG:
            print(f"RecentInfo data = {data}")
        return data

    def getPartOfTheDay(self):
        currentHour = int(datetime.today().strftime("%H"))
        if currentHour == 0 :
            return "Midnight"
        elif currentHour >= 1  and currentHour < 5:
            return "Late Night"
        elif currentHour >= 5 and currentHour < 8:
            return "Early Morning"
        elif currentHour >= 8 and currentHour < 12:
            return "Morning"
        elif currentHour == 12:
            return "Noon"
        elif currentHour >= 13 and currentHour < 18:
            return "Afternoon"
        elif currentHour >= 18 and currentHour < 24:
            return "Evening"

    def greeter(self):
        partOfDay = self.getPartOfTheDay()
        if  partOfDay == "Evening":
            return "Evening"
        elif partOfDay in ["Midnight", "Early Morning", "Morning", "Late Night"]:
            return "Morning"
        elif partOfDay in ["Noon","Afternoon"]:
            return "Afternoon"
        else:
            return "Error"

    def inputNewName(self):
        isRunning = True
        data = ""
        while isRunning:
            with output.use_tags('inputNewName'):
                data = UI.set_userInput("New Username: ",f"{'Welcome New User, Please Input your name.' if not self.getUsername() else 'Hi '+ self.getUsername() +' what would your new name be like?'}")
                if UI.set_verifyInputBool(f"{'Are you sure you name is '+ data if not self.getUsername() else 'Are you sure you want to change your name as '+ self.getUsername() +'?'}"):
                    UI.display_msg("success",f"Successfully {'set'  if not self.getUsername() else 'changed'} your Username, Welcome {self.getUsername()}!")
                    self.setUsername(data.capitalize())
                    isRunning = False
                    continue
                if self.getUsername() and UI.set_verifyInputBool(f"Continue changing your name? {self.getUsername()}"):
                    isRunning = False
            output.clear(output_tags='inputNewName')  


    def updateInfo(self):
        data = self.getRecentInfo()
        if not data:
            with output.use_tags('nameInit'):
                UI.display_msg("warning","No recent Userdata found. Now Creating..")
                self.inputNewName()
            output.clear(output_tags='nameInit')
            UI.display_msg("success","User Initialization Done..")
        else:
            self.setUsername(data[0]) 
        self.updateDownloadCounter()
        self.saveRecentInfo()

    def greetUser(self):
        self.updateInfo()
        userStats = [f"Total Torrent Downloads: {self._totalDownloads}",f"Total Unfinished Downloads: {len(File_Manager.get_Unfinished())}",f"{GeneralFunc.getDaysTimePast({self._lasts},GeneralFunc.getCurrentDate())}"]
        print(f"\n{color['yellow']}Welcome and Good {self.greeter()}! {color['bold']}{color['green']}{self.getUsername()}{color['end']}.")
        print(f"{color['yellow']}Have a great {self.getPartOfTheDay()}!{color['end']}")
        print(f"\n{color['bold']}{color['yellow']}Stats:{color['end']}")
        UI.display_items(userStats)


        
x = Personalization()
x.greetUser()





```
